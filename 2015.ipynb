# --- Day 7: Some Assembly Required ---

```python
puzzle = '''123 -> x
456 -> y
x AND y -> d
x OR y -> e
x LSHIFT 2 -> f
y RSHIFT 2 -> g
NOT x -> h
NOT y -> i'''
with open('7.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle

N = len(puzzle)

################################################################################
# PART 1

# processing
wires = {}
for line in puzzle:

    assert not line[-1].isnumeric(), 'TARGET REGISTER MUST BE STRING'
    w = line[-1]

    if len(line) == 3: # x -> y
        wires[w] = { 'op' : 'ASSIGN', 'args' : [ int(line[0]) if line[0].isnumeric() else line[0] ] }
    elif len(line) == 4: # NOT x -> y
        wires[w] = { 'op' : 'NOT', 'args' : [ int(line[1]) if line[1].isnumeric() else line[1] ] }
    else: # x OR / AND / LSHIFT / RSHIFT y -> y
        assert len(line) == 5
        assert line[1] in ['AND', 'OR', 'LSHIFT', 'RSHIFT'], 'UNRECOGNIZED OP'
        wires[w] = { 'op' : line[1], 'args' : [ int(line[0]) if line[0].isnumeric() else line[0], int(line[2]) if line[2].isnumeric() else line[2] ] }

assert len(wires) == N
wires


# layering wires - 1st layer does not depend on others
for w in wires:
    depends_on_other_wires = any(a in wires for a in wires[w]['args'])
    if not depends_on_other_wires:
        wires[w]['layer'] = 0
    else:
        wires[w]['layer'] = None

{ w : wires[w] for w in wires if wires[w]['layer'] == 0 }


# layering wires - additional layers built upon each other
l = 1
while True:

    parent_wires = [ w for w in wires if wires[w]['layer'] != None ]
    # print('Wires already done:', parent_wires)
    
    for w in wires:
        if wires[w]['layer'] == None:
            args = [ a for a in wires[w]['args'] if type(a) == str ]
            all_args_ready = all( a in parent_wires for a in args )
            if all_args_ready:
                # print('Wire', w, 'is ready.')
                wires[w]['layer'] = l
    
    
    all_layers_finished = all( wires[w]['layer'] != None for w in wires )
    if all_layers_finished: break

    l += 1

wires


def run_wires(wires):

    L = max(wires[w]['layer'] for w in wires)
    for l in range(L + 1):

        ready_wires = [ w for w in wires if wires[w]['layer'] == l ]
        for w in ready_wires:
            op = wires[w]['op']

            if op in ['ASSIGN', 'NOT']:
                assert len(wires[w]['args']) == 1, 'ARGUMENTS ERROR'

                a = wires[w]['args'][0]
                a = a if type(a) == int else wires[a]['value']

                if op == 'ASSIGN':
                    wires[w]['value'] = a
                else:
                    wires[w]['value'] = ~a % (2 ** 16)

            elif op in ['AND', 'OR', 'LSHIFT', 'RSHIFT']:
                assert len(wires[w]['args']) == 2, 'ARGUMENTS ERROR'

                a0 = wires[w]['args'][0]
                a0 = a0 if type(a0) == int else wires[a0]['value']
                a1 = wires[w]['args'][1]
                a1 = a1 if type(a1) == int else wires[a1]['value']

                if op == 'AND':
                    wires[w]['value'] = a0 & a1
                elif op == 'OR':
                    wires[w]['value'] = a0 | a1
                elif op == 'LSHIFT':
                    wires[w]['value'] = a0 << a1
                else:
                    wires[w]['value'] = a0 >> a1

    return wires

wires = run_wires(wires)
wires['a']

################################################################################
# PART 1


a = wires['a']['value']
wires['b']['args'] = [ a ]

wires = run_wires(wires)
wires['a']
```

# --- Day 9: All in a Single Night ---

```python
puzzle = '''London to Dublin = 464
London to Belfast = 518
Dublin to Belfast = 141'''
with open('9.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ route.split(' ') for route in puzzle ]
puzzle = [ [ route[0], route[2], int(route[4]) ] for route in puzzle ]
puzzle

################################################################################
# PART 1-2

distances = {}
for route in puzzle:
    
    start = route[0]
    target = route[1]
    dist = route[2]

    if start not in distances:
        distances[start] = { target : dist }
    else:
        distances[start][target] = dist

    if target not in distances:
        distances[target] = { start : dist }
    else:
        distances[target][start] = dist

distances
N = len(distances)


#
from itertools import permutations
perms = permutations(distances)

MIN = 10 ** 6
MAX = 0
for p in perms:

    total = sum( distances[ p[i - 1] ][ p[i] ] for i in range(1, N) )
    
    if total < MIN: MIN = total
    if MAX < total: MAX = total

MIN, MAX
```

# --- Day 10: Elves Look, Elves Say ---

```python
puzzle = '1'
puzzle = '11'
puzzle = '21'
puzzle = '1211'
puzzle = '111221'

puzzle = '1113222113' # puzzle input

puzzle = list(puzzle)
puzzle = list(map(int, puzzle))
puzzle

################################################################################
# PART 1-2

def look_and_say(look):
    N = len(look)

    say = []
    i = 0
    while i < N:

        j = i + 1
        while j < N:
            if look[i] != look[j]:
                break
            j += 1

        if j != N:
            tmp = look[i : j]
            i = j
        else:
            tmp = look[i : ]
            i = N

        say.append(len(tmp))
        say.append(tmp[0])
    
    return say


# PART 1: 40 times
result = puzzle
for i in range(40): result = look_and_say(result)
len(result)


# PART 2: 50 times = 10 more times
for i in range(10): result = look_and_say(result)
len(result)
```

# --- Day 11: Corporate Policy ---

```python
puzzle = 'abcdefgh'
puzzle = 'ghijklmn'
puzzle = 'hepxcrrq' # puzzle input

puzzle

################################################################################
# PART 1-2

def find_next_password(current):
    abc = 'abcdefghijklmnopqrstuvwxyz'

    indices = [ abc.index(char) for char in current ] # preprocessing if current password contains ilo
    i = indices.index(8) if 8 in indices else 8
    l = indices.index(11) if 11 in indices else 8
    o = indices.index(14) if 14 in indices else 8
    m = min(i, l, o) # find first occurence of bad index
    if m != 8: # increase index and replace everything afterwards
        indices[m] += 1 
        indices[ m + 1 : ] = [0] * (7 - m)


    while True:

        # increase letter indices
        i = 7
        while i >= 0:
            indices[i] += 1
            if indices[i] in [8, 11, 14]: # forbidden i, l, o letter
                indices[i] += 1
            if indices[i] == 26: # wrapping around
                indices[i] = 0
                i -= 1
            else:
                break

        # 3 subsequent letters
        trio = False
        for i in range(6):
            trio = (indices[i] + 1 == indices[i + 1] and indices[i + 1] + 1 == indices[i + 2])
            if trio: break

        # 2 distinct pairs of letters
        duo = False
        pairs = [ tuple(indices[i : i + 2]) for i in range(7) if indices[i] == indices[i + 1] ]
        pairs = set(pairs)
        duo = ( len(pairs) >= 2 )

        # code found
        if duo and trio:
            break

    new_pw = ''.join([ abc[i] for i in indices ])
    return new_pw

# PART 1
pw = find_next_password(puzzle)
pw

# PART 2
find_next_password(pw)
```

# --- Day 13: Knights of the Dinner Table ---

```python
puzzle = '''Alice would gain 54 happiness units by sitting next to Bob.
Alice would lose 79 happiness units by sitting next to Carol.
Alice would lose 2 happiness units by sitting next to David.
Bob would gain 83 happiness units by sitting next to Alice.
Bob would lose 7 happiness units by sitting next to Carol.
Bob would lose 63 happiness units by sitting next to David.
Carol would lose 62 happiness units by sitting next to Alice.
Carol would gain 60 happiness units by sitting next to Bob.
Carol would gain 55 happiness units by sitting next to David.
David would gain 46 happiness units by sitting next to Alice.
David would lose 7 happiness units by sitting next to Bob.
David would gain 41 happiness units by sitting next to Carol.'''
with open('13.txt') as f: puzzle = f.read()


# processing
puzzle = puzzle.replace('.', '').split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle = [ [ line[0], line[2], line[3], line[-1] ] for line in puzzle ]
puzzle = [ [ line[0], (1 if line[1] == 'gain' else -1) * int(line[2]), line[3] ] for line in puzzle ]
puzzle

################################################################################
# PART 1

sitting = {}
for line in puzzle:
    name0 = line[0]
    name1 = line[2]
    s = line[1]

    if name0 not in sitting:
        sitting[ name0 ] = { name1 : s }
    else:
        sitting[ name0 ][ name1 ] = s

sitting


#
from itertools import permutations

def find_best_sitting(sitting):
    N = len(sitting)
    perms = permutations(sitting)

    M = 0
    for p in perms:
        m = sum( sitting[ p[i - 1] ][ p[i] ] + sitting[ p[i] ][ p[i - 1] ] for i in range(N) )
        if M < m: M = m

    return M

find_best_sitting(sitting)

################################################################################
# PART 2

# add myself to the table
for s in sitting: sitting[s]['Me'] = 0
sitting['Me'] = { s : 0 for s in sitting }
sitting

find_best_sitting(sitting)
```

# --- Day 14: Reindeer Olympics ---

```python
puzzle = '''Comet can fly 14 km/s for 10 seconds, but then must rest for 127 seconds.
Dancer can fly 16 km/s for 11 seconds, but then must rest for 162 seconds.'''
with open('14.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ reindeer.split(' ') for reindeer in puzzle ]
puzzle = [ [ r[0], r[3], r[6], r[-2] ] for r in puzzle ]
puzzle

################################################################################
# PART 1

reindeers = { r[0] : {'speed' : int(r[1]), 'stamina' : int(r[2]), 'rest' : int(r[3]), 'dist' : 0 } for r in puzzle }
reindeers

reindeer_statuses = { r : { 'dist' : 0, 'rest' : False, 'time' : reindeers[r]['stamina'] } for r in reindeers }
reindeer_statuses

for i in range(2503):

    for r in reindeers:
        if not reindeer_statuses[r]['rest']: # reindeer r is not resting, then move
            reindeer_statuses[r]['dist'] += reindeers[r]['speed']
            reindeer_statuses[r]['time'] -= 1
            if reindeer_statuses[r]['time'] == 0: # stamina timer depletes => switch to resting
                reindeer_statuses[r]['rest'] = True
                reindeer_statuses[r]['time'] = reindeers[r]['rest']
        else:
            reindeer_statuses[r]['time'] -= 1
            if reindeer_statuses[r]['time'] == 0: # resting timer ends => switch to moving
                reindeer_statuses[r]['rest'] = False
                reindeer_statuses[r]['time'] = reindeers[r]['stamina']

max( reindeer_statuses[r]['dist'] for r in reindeer_statuses )

################################################################################
# PART 2

reindeer_statuses = { r : { 'dist' : 0, 'rest' : False, 'time' : reindeers[r]['stamina'], 'score' : 0 } for r in reindeers } # status reset + scoring
reindeer_statuses

for i in range(2503): # same code + scoring

    for r in reindeers:
        if not reindeer_statuses[r]['rest']: # reindeer r is not resting, then move
            reindeer_statuses[r]['dist'] += reindeers[r]['speed']
            reindeer_statuses[r]['time'] -= 1
            if reindeer_statuses[r]['time'] == 0: # stamina timer depletes => switch to resting
                reindeer_statuses[r]['rest'] = True
                reindeer_statuses[r]['time'] = reindeers[r]['rest']
        else:
            reindeer_statuses[r]['time'] -= 1
            if reindeer_statuses[r]['time'] == 0: # resting timer ends => switch to moving
                reindeer_statuses[r]['rest'] = False
                reindeer_statuses[r]['time'] = reindeers[r]['stamina']

    # scoring
    R, _ = max( [ (r, reindeer_statuses[r]['dist']) for r in reindeer_statuses ] , key = lambda x: x[1] )
    reindeer_statuses[R]['score'] += 1

max( reindeer_statuses[r]['score'] for r in reindeer_statuses )
```

#--- Day 15: Science for Hungry People ---

```python
puzzle = '''Butterscotch: capacity -1, durability -2, flavor 6, texture 3, calories 8
Cinnamon: capacity 2, durability 3, flavor -2, texture -1, calories 3'''
with open('15.txt') as f: puzzle = f.read()

puzzle = puzzle.replace(':', '').replace(',', '')
puzzle = puzzle.split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle = [ [ int(ingredient[2]), int(ingredient[4]), int(ingredient[6]), int(ingredient[8]), int(ingredient[10]) ] for ingredient in puzzle ]
puzzle

################################################################################
# PART 1-2

limit = 100
weights = []

# grid search
def calculate_weights(w = [], d = 1):
    global puzzle, weights
    s = sum(w)

    if d == len(puzzle):
        w.append( limit - s )
        weights.append(w)
    else:
        for i in range(0, limit + 1 - s):
            calculate_weights(w + [i], d + 1)

calculate_weights()
weights


#
S = 0
for w in weights:

    capacity = 0
    durability = 0
    flavor = 0
    texture = 0
    calories = 0

    for i in range(len(puzzle)):
        x = puzzle[i]

        capacity += x[0] * w[i]
        durability += x[1] * w[i]
        flavor += x[2] * w[i]
        texture += x[3] * w[i]
        calories += x[4] * w[i]

    capacity = max(capacity, 0)
    durability = max(durability, 0)
    flavor = max(flavor, 0)
    texture = max(texture, 0)
    assert calories > 0

    s = capacity * durability * flavor * texture

    # # part 1 simple max search
    # if S < s:
    #     S = s

    # part 2 with calories = 500 condition
    if calories == 500:
        if S < s: S = s

S
```

```python
puzzle = '''.#.#.#
...##.
#....#
..#...
#.#..#
####..'''
with open('18.txt') as f: puzzle = f.read()

import numpy as np
puzzle = puzzle.split('\n')
puzzle = [ list(line) for line in puzzle ]
puzzle = np.array(puzzle)
puzzle

################################################################################
# PART 1-2

R, C = puzzle.shape
current = puzzle

# part 2 - corners cannot be turned off
current[0, 0] = '#'
current[0, C - 1] = '#'
current[R - 1, 0] = '#'
current[R - 1, C - 1] = '#'


for iter in range(100):
    next = np.empty(shape = (R, C), dtype = '<U1')

    for r in range(R):
        for c in range(C):

            coords = []
            for i in [-1, 0, 1]:
                for j in [-1, 0, 1]:
                    if 0 <= r + i and r + i < R and 0 <= c + j and c + j < C:
                        if i != 0 or j != 0:
                            coords.append([r + i, c + j])
            coords = np.array(coords)

            neighbours = current[coords[:, 0], coords[:, 1]]
            neighbours = np.sum(neighbours == '#')

            if current[r, c] == '#':
                if neighbours in [2, 3]:
                    next[r, c] = '#'
                else:
                    next[r, c] = '.'
            else:
                if neighbours == 3:
                    next[r, c] = '#'
                else:
                    next[r, c] = '.'

            # part 2 - corners cannot be turned off
            next[0, 0] = '#'
            next[0, C - 1] = '#'
            next[R - 1, 0] = '#'
            next[R - 1, C - 1] = '#'

    current = next

np.sum(current == '#')
```

# --- Day 20: Infinite Elves and Infinite Houses ---

```python
puzzle = 34 * (10 ** 6)

################################################################################
# PART 1

from math import sqrt

def sum_divisors(n):
    divs = [1, n]
    for i in range(2, int(sqrt(n)) + 1): # iterate only to sqrt of n for efficiency
        if n % i == 0:
            divs.append(i)
            divs.append(n // i)
    return sum(divs)

n = 1
while sum_divisors(n) * 10 < puzzle:
    n += 1

n

################################################################################
# PART 2

houses = {}

n = 1
while True:

    houses[n] = 0

    for d in range(1, int(sqrt(n)) + 1): # iterate only to sqrt of n for efficiency
        if n % d == 0:

            # divisor d is on the "lower" end, hence n // d is on the "higher"
            # deliver presents equal to eleven times their number at each house
            if d <= 50: houses[n] += (n // d * 11)
            if n // d <= 50: houses[n] += (d * 11)
            
    if houses[n] > puzzle: break
    n += 1

n
```

```python
puzzle = '''inc a
jio a, +2
tpl a
inc a'''
with open('23.txt') as f: puzzle = f.read()

puzzle = puzzle.replace(',', '')
puzzle = puzzle.split('\n')
puzzle = [ cmd.split(' ') for cmd in puzzle ]
puzzle

################################################################################
# PART 1-2

registers = { 'a' : 0, 'b' : 0 } # PART 1
registers = { 'a' : 1, 'b' : 0 } # PART 2

i = 0
while True:

    if i < 0 or i >= len(puzzle):
        break

    cmd = puzzle[i]

    if cmd[0] == 'hlf':
        r = cmd[1]
        registers[r] /= 2
        registers[r] = int(registers[r])
    if cmd[0] == 'tpl':
        r = cmd[1]
        registers[r] *= 3
    if cmd[0] == 'inc':
        r = cmd[1]
        registers[r] += 1
    if cmd[0] == 'jmp':
        v = cmd[1]
        v = registers[v] if v in registers else int(v)
        i += v
        continue
    if cmd[0] == 'jie':
        r = cmd[1]
        if registers[r] % 2 == 0:
            v = int(cmd[2])
            i += v
            continue
    if cmd[0] == 'jio':
        r = cmd[1]
        if registers[r] == 1:
            v = int(cmd[2])
            i += v
            continue

    i += 1

registers['b']
```
