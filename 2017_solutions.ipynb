# --- Day 3: Spiral Memory ---

```python
puzzle = 312051

################################################################################
# PART 1

# find closest odd square number which is greater than the target (= this is the bottom right of the box containing the spiral)
l = 1
while l * l < puzzle:
    l += 2

import numpy as np
mem = np.zeros(shape = (l, l), dtype = int)
c = l // 2
mem[c, c] = 1

def rotate_90(dir):
    return np.array([ -dir[1], dir[0] ])



# move counter-clockwise on the grid, starting on the right of the center + facing east
idx = np.array([c, c])
dir = np.array([0, 1])

# in each iteration
for i in range(2, puzzle + 1):
    # move to a new pos using facing direction, then fill the blank 0 with a number
    idx += dir
    assert mem[ idx[0], idx[1] ] == 0, 'Writing to memory index with value already on it!'
    mem[ idx[0] , idx[1] ] = i

    # facing east = if have nothing above, then change facing to up
    if dir[1] == 1:
        if mem[ idx[0] - 1, idx[1] ] == 0:
            dir = rotate_90(dir)
        continue

    # facing up = if have nothing left, then change facing to left
    if dir[0] == -1:
        if mem[ idx[0] , idx[1] - 1 ] == 0:
            dir = rotate_90(dir)
        continue

    # similar logic here
    if dir[1] == -1:
        if mem[ idx[0] + 1 , idx[1] ] == 0:
            dir = rotate_90(dir)
        continue

    # and here
    if dir[0] == 1:
        if mem[ idx[0] , idx[1] + 1 ] == 0:
            dir = rotate_90(dir)
        continue


# find puzzle value, then calculate manhattan distance from center
idx = np.where(mem == puzzle)
idx = np.array([ idx[0][0], idx[1][0] ])
idx -= c

np.sum(np.abs(idx))


################################################################################
# PART 2

# to check neighbouring values, we need a one step bigger grid
mem = np.zeros(shape = (1 + l + 1, 1 + l + 1), dtype = int)
mem[ c + 1 ][ c + 1 ] = 1

# following the very same logic as before
idx = np.array([c + 1, c + 1])
dir = np.array([0, 1])

while True:
    idx += dir
    assert mem[ idx[0], idx[1] ] == 0, 'Writing to memory index with value already on it!'

    # changing only the filling logic
    v = 0
    for y in [-1, 0, 1]:
        for x in [-1, 0, 1]:
            v += mem[ idx[0] + y , idx[1] + x ]
    mem[ idx[0] , idx[1] ] = v

    # break out and get the answer
    if v > puzzle:
        break


    if dir[1] == 1:
        if mem[ idx[0] - 1, idx[1] ] == 0:
            dir = rotate_90(dir)
        continue
    if dir[0] == -1:
        if mem[ idx[0] , idx[1] - 1 ] == 0:
            dir = rotate_90(dir)
        continue
    if dir[1] == -1:
        if mem[ idx[0] + 1 , idx[1] ] == 0:
            dir = rotate_90(dir)
        continue
    if dir[0] == 1:
        if mem[ idx[0] , idx[1] + 1 ] == 0:
            dir = rotate_90(dir)
        continue

v
```

# --- Day 6: Memory Reallocation ---

```python
# example
puzzle = '0,2,7,0'
with open('6.txt') as f: puzzle = f.read()

puzzle = puzzle.replace('\t', ',') # transform the puzzle input
puzzle = puzzle.split(',')
puzzle = list(map(int, puzzle))
puzzle

################################################################################
# PART 1

states = []
states.append(tuple(puzzle))

s = states[0]
iter = 0
while True:
    iter += 1
    s = list(s)
    
    # find and distribute max
    m = max(s)
    i = s.index(m)
    s[i] = 0
    for j in range(m):
        i += 1
        i = i % len(s)
        s[i] += 1

    s = tuple(s) # need back & forth conversion to avoid references
    if s not in states:
        states.append(s)
    else:
        # exiting when encountering state loop
        break

iter

################################################################################
# PART 2

# states are added sequentially to the end, so the loop size must be the total number of states minus the index of the loop start
len(states) - states.index(s)
```

# --- Day 7: Recursive Circus ---

```python
puzzle = '''pbga (66)
xhth (57)
ebii (61)
havc (66)
ktlj (57)
fwft (72) -> ktlj, cntj, xhth
qoyq (66)
padx (45) -> pbga, havc, qoyq
tknk (41) -> ugml, padx, fwft
jptl (61)
ugml (68) -> gyxo, ebii, jptl
gyxo (61)
cntj (57)'''
with open('7.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, line in enumerate(puzzle):
    program = line.split(' ')[0]

    self_weight = line.split(' ')[1]
    self_weight = int(self_weight[1 : -1])

    new_line = [program, self_weight]

    idx = line.find('>')
    if idx != -1:
        idx += 2
        children = line[idx :].split(', ')
        new_line.append(children)

    puzzle[i] = new_line

puzzle = { line[0] : { 'self_weight' : line[1] } if len(line) == 2 else { 'self_weight' : line[1], 'children' : line[2] } for line in puzzle } # this is a lil' bit ugly
puzzle

################################################################################
# PART 1

bottom = list(puzzle)[0]

while True:

    has_parent = False
    for program in puzzle:
        if 'children' in puzzle[program]:
            if bottom in puzzle[program]['children']:
                has_parent = True
                break

    if has_parent:
        bottom = program
    else:
        break

bottom

################################################################################
# PART 2

# calculate the total weight of programs discs recursively
def calc_total_weight(program):
    global puzzle

    if 'children' not in puzzle[program]:
        puzzle[program]['total_weight'] = puzzle[program]['self_weight']
        return puzzle[program]['self_weight']
    else:
        tw = puzzle[program]['self_weight']
        for c in puzzle[program]['children']:
            tw += calc_total_weight(c)
    
    puzzle[program]['total_weight'] = tw
    return tw

calc_total_weight(bottom)
puzzle


# find program with weight mismatch
for program in puzzle:
    if 'children' in puzzle[program]:
        weights = { c : puzzle[c]['total_weight'] for c in puzzle[program]['children'] }
        if len(set(weights.values())) != 1:
            break

# identify bad weight and its program
w = sorted(weights.values())
w = w[2] if w[0] == w[1] else w[0]
p = [ p for p in weights if weights[p] == w ][0]
p, w

# answer = add the weight difference to self weight
puzzle[p]['self_weight'] + (list(weights.values())[1] - w)
```

# --- Day 8: I Heard You Like Registers ---

```python
puzzle = '''b inc 5 if a > 1
a inc 1 if b < 5
c dec -10 if a >= 1
c inc -20 if c == 10'''
with open('8.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, line in enumerate(puzzle):
    line = line.split(' ')
    line = line[: 3] + line[4 :]

    line[2] = int(line[2])
    if line[1] == 'dec': line[2] *= -1 # remove inc / dec by multiplying with -1 if needed (= all instructions now add)
    del line[1]

    line[-1] = int(line[-1])

    puzzle[i] = line

puzzle

################################################################################
# PART 1-2

registers = {}

for instruction in puzzle:
    r = instruction[0]
    if r not in registers: registers[ r ] = 0
    r = instruction[2]
    if r not in registers: registers[ r ] = 0

registers


m = 0
for instruction in puzzle: # execute the instructions

    r1 = instruction[0]
    v1 = instruction[1]
    r2 = instruction[2]
    sym = instruction[3]
    v2 = instruction[4]

    condition = False
    if sym == '==': condition = (registers[ r2 ] == v2)
    if sym == '!=': condition = (registers[ r2 ] != v2)
    if sym == '>': condition = (registers[ r2 ] > v2)
    if sym == '<': condition = (registers[ r2 ] < v2)
    if sym == '>=': condition = (registers[ r2 ] >= v2)
    if sym == '<=': condition = (registers[ r2 ] <= v2)
        
    if condition: registers[ r1 ] += v1

    # keep track of running maximum for part 2
    if m < max(registers.values()):
        m = max(registers.values())

max(registers.values()), m
```

```python
puzzle = '3,4,1,5'
n = 5

with open('10.txt') as f:
    puzzle = f.read()
    n = 256

puzzle = puzzle.split(',')
puzzle = list(map(int, puzzle))
nums = list(range(n))
nums, puzzle

################################################################################
# PART 1

def knotting(pos = 0, skip = 0):
    global puzzle, nums

    # in each iteration: select region of interest, revert, then insert back
    for length in puzzle:

        # region of interest fits in list "normally"
        if pos + length <= n:
            tmp = nums[pos : pos + length]
            tmp = tmp[ : : -1]
            nums = nums[: pos] + tmp + nums[pos + length :]

        # it does not :'(
        else:
            suffix = nums[pos :]
            r = pos + length - n
            prefix = nums[: r] # r == len(prefix)
            tmp = suffix + prefix
            assert len(tmp) == length, 'Length error!'

            tmp = tmp[: : -1]
            suffix = tmp[: len(suffix)]
            prefix = tmp[len(suffix) :]
            nums = prefix + nums[r : pos] + suffix

        # change pos, increase skip size
        pos += (length + skip)
        pos %= n
        skip += 1

    return pos, skip

pos, skip = knotting()
nums[0] * nums[1]

################################################################################
# PART 2

# puzzle, nums reset
puzzle = '' # a2582a3a0e66e6e86e3812dcb672a272
puzzle = 'AoC 2017' #  33efeb34ea91902bb2f59c9920caa6cd.
puzzle = '1,2,3' #  3efbe78a8d82f29979031a4aa0b16a9d.
puzzle = '1,2,4' # 63960835bcdc130f0b66d7ff4f6a5a8e
with open('10.txt') as f: puzzle = f.read()
n = 256
nums = list(range(n))


puzzle = [ ord(char) for char in puzzle ] # convert to ASCII
puzzle += [17, 31, 73, 47, 23] # rule from the problem, have to add


# run 64 knots
pos = 0
skip = 0
for iter in range(64):
    pos, skip = knotting(pos, skip)


# convert to hash
h = ''
for i in range(0, 256, 16):
    l = nums[i : i + 16]
    
    xor = l[0]
    for j in range(1, 16):
        xor = xor ^ l[j]

    xor = hex(xor)[2:]
    if len(xor) == 1: xor = '0' + xor # add leading 0 in case of 1-digit hex
    h += xor

h
```

# --- Day 12: Digital Plumber ---

```python
puzzle = '''0 <-> 2
1 <-> 1
2 <-> 0, 3, 4
3 <-> 2, 4
4 <-> 2, 3, 6
5 <-> 6
6 <-> 4, 5'''
with open('12.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ program.replace(',', '').split(' ') for program in puzzle ]
puzzle = [ program[2 :] for program in puzzle ]
puzzle = [ list(map(int, program)) for program in puzzle ]
puzzle

################################################################################
# PART 1

def programs_connected(main):
    global puzzle

    done = [] # return list of programs forming a connected component with main
    
    current = [main] # starting from program main
    while True:
        # find those programs which are connected to the current (breadt search schema)
        next = []

        for p in current: # pick one program from curent checklist
            for pp in puzzle[p]:

                # add a new program to check if
                # 1. it's not already done
                # 2. it's not in current checklist
                # 3. it's not added to the next iter yet
                if (pp not in done) and (pp not in current) and (pp not in next):
                    next.append(pp)

            done.append(p) # mark current program done
        
        # break out if no next batch, else copy and start again
        if next == []:
            break
        else:
            current = next.copy()

    return done

len(programs_connected(0))

################################################################################
# PART 2

N = len(puzzle)
programs = list(range(N))

groups = []
while True:
    # find all programs connected to main (= forming a group)
    main = min(programs)
    g = programs_connected(main)

    # separate the group, decrease iter
    groups.append(g)
    N -= len(g)
    programs = [ p for p in programs if p not in g ]
    
    if N == 0: break

len(groups)
```

# --- Day 13: Packet Scanners ---

```python
puzzle = '''0: 3
1: 2
4: 4
6: 4'''
with open('13.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ layer.split(': ') for layer in puzzle ]
puzzle = [ list(map(int, layer)) for layer in puzzle ]
puzzle = { layer[0] : { 'max' : layer[1], 'pos' : 0, 'dir' : 1 } for layer in puzzle }
puzzle

################################################################################
# PART 1

severity = 0

current_layer = -1
last = max(puzzle)

while current_layer != last:

    # stepping into the layer
    current_layer += 1
    if current_layer in puzzle: # if scanner is on pos 0, then caught => increase severity
        scanner = puzzle[current_layer]
        if scanner['pos'] == 0:
            severity += scanner['max'] * current_layer

    # move scanners afterwards
    for layer in puzzle:
        scanner = puzzle[layer]
        scanner['pos'] += scanner['dir']
        if scanner['pos'] == scanner['max'] - 1:
            scanner['dir'] = -1
        if scanner['pos'] == 0:
            scanner['dir'] = 1

severity

################################################################################
# PART 2

# Cycles = 2 * (Range - 1) -- the amount of the to make a full trip from and back to 0

# | Layer | Range | Cycles | Statement        | Reduces to   |
# |-------|-------|--------|------------------|--------------|
# |     0 |     3 |      4 | T + 0 != 0 mod 4 | T != 0 mod 4 |
# |     1 |     2 |      2 | T + 1 != 0 mod 2 | T != 1 mod 2 |
# |     4 |     4 |      6 | T + 4 != 0 mod 6 | T != 2 mod 6 |
# |     6 |     4 |      6 | T + 6 != 0 mod 6 | T != 0 mod 6 |

# idea to reduce layers + ranges into a sequence of constrained modular equations: for each modulo 'M', the solution cannot have remainder 'R' mod 'M'

# save cycle size, table attached: f(n) = 2 * n - 2
for layer in puzzle: puzzle[layer]['cyc'] = 2 * puzzle[layer]['max'] - 2

moduli = {}
for layer in puzzle:
    scanner = puzzle[layer]

    m = scanner['cyc']
    r = m - (layer % m)
    r %= m

    if m not in moduli:
        moduli[m] = [r]
    else:
        if r not in moduli[m]:
            moduli[m].append(r)

moduli

# brute force search for the interval containing at least one solution to the constraints (increasing search range 10-fold)
i = 0
while True:

    solutions = list( range(10 ** i, 10 ** (i + 1)) )
    for m in moduli: solutions = [ s for s in solutions if s % m not in moduli[m] ]
                                  
    if solutions != []:
        break
    else:
        i += 1

min(solutions)
```

# --- Day 16: Permutation Promenade ---

```python
puzzle = '''s1,x3/4,pe/b'''
abc = list('abcde')

with open('16.txt') as f:
    puzzle = f.read()
    abc = list('abcdefghijklmnop')

puzzle = puzzle.split(',')
puzzle

N = len(abc)

################################################################################
# PART 1

def dance(abc):
    global puzzle

    for line in puzzle:
        
        if line[0] == 's':
            idx = int(line[1:])
            abc = abc[-idx : ] + abc[: -idx]
            continue

        if line[0] == 'x':
            pos = line[1:].split('/')
            pos = list(map(int, pos))
            tmp = abc[ pos[0] ]
            abc[ pos[0] ] = abc[ pos[1] ]
            abc[ pos[1] ] = tmp
            continue

        if line[0] == 'p':
            prog = line[1:].split('/')
            i = abc.index(prog[0])
            j = abc.index(prog[1])
            abc[i] = prog[1]
            abc[j] = prog[0]
            continue

    return abc

abc = dance(abc)
''.join(abc)

################################################################################
# PART 2

from math import factorial as fact

# reset the abc, and while looping - if encountering the starting pos, then we can find the 10**9th iter by taking mod cycle size
abc = sorted(abc)
for i in range(fact(N)):
    abc = dance(abc)
    if abc == sorted(abc): break


cycle = i + 1
abc = sorted(abc)
for i in range((10 ** 9) % cycle):
    abc = dance(abc)
''.join(abc)
```

# --- Day 18: Duet ---

```python
puzzle = '''set a 1
add a 2
mul a a
mod a 5
snd a
set a 0
rcv a
jgz a -1
set a 1
jgz a -2'''
puzzle = '''snd 1
snd 2
snd p
rcv a
rcv b
rcv c
rcv d'''
with open('18.txt') as f: puzzle = f.read()

abc = 'abcdefghijklomnopqrstuvwxyz'

puzzle = puzzle.split('\n')
puzzle = [ cmd.split(' ') for cmd in puzzle ]
puzzle

################################################################################
# PART 1

registers = {}

for cmd in puzzle:
    if (cmd[1] in abc) and (cmd[1] not in registers):
        registers[ cmd[1] ] = 0
    
registers


i = 0
sounds = []
while True:
    cmd = puzzle[i]

    if cmd[0] == 'snd':
        val = cmd[1]
        val = registers[val] if val in registers else int(val)
        sounds.append(val)

    if cmd[0] == 'set':
        val = cmd[2]
        val = registers[val] if val in registers else int(val)
        registers[ cmd[1] ] = val
        
    if cmd[0] == 'add':
        val = cmd[2]
        val = registers[val] if val in registers else int(val)
        registers[ cmd[1] ] += val
    
    if cmd[0] == 'mul':
        val = cmd[2]
        val = registers[val] if val in registers else int(val)
        registers[ cmd[1] ] *= val
    
    if cmd[0] == 'mod':
        val = cmd[2]
        val = registers[val] if val in registers else int(val)
        registers[ cmd[1] ] %= val

    if cmd[0] == 'rcv':
        reg = cmd[1]
        if registers[reg] != 0:
            break

    if cmd[0] == 'jgz':
        val = cmd[1]
        val = registers[val] if val in registers else int(val)
        if val > 0:
            val = cmd[2]
            val = registers[val] if val in registers else int(val)
            i += val
            continue

    i += 1

sounds[-1]

################################################################################
# PART 2

# register reset, and 'p' register init
for r in registers: registers[r] = 0
registers = [registers.copy(), registers.copy()]
registers[1]['p'] = 1
registers

# init positions, start running program 1
positions = [0, 0]
sounds = [[], []]
p = 0

c = 0 # counter how many times program 1 makes a sound

# same code, but added register indexing
while True:
    
    cmd = puzzle[ positions[p] ]
    # print('Program:', p, '; position:', positions[p], 'command:', cmd)

    if cmd[0] == 'snd':
        val = cmd[1]
        val = registers[p][val] if val in registers[p] else int(val)
        sounds[1 - p].append(val)
        c += (p == 1)
    if cmd[0] == 'set':
        val = cmd[2]
        val = registers[p][val] if val in registers[p] else int(val)
        registers[p][ cmd[1] ] = val
    if cmd[0] == 'add':
        val = cmd[2]
        val = registers[p][val] if val in registers[p] else int(val)
        registers[p][ cmd[1] ] += val
    if cmd[0] == 'mul':
        val = cmd[2]
        val = registers[p][val] if val in registers[p] else int(val)
        registers[p][ cmd[1] ] *= val
    if cmd[0] == 'mod':
        val = cmd[2]
        val = registers[p][val] if val in registers[p] else int(val)
        registers[p][ cmd[1] ] %= val
    if cmd[0] == 'jgz':
        val = cmd[1]
        val = registers[p][val] if val in registers[p] else int(val)
        if val > 0:
            val = cmd[2]
            val = registers[p][val] if val in registers[p] else int(val)
            positions[p] += val
            continue

    # modifying the receive command handling
    if cmd[0] == 'rcv':

        # has sound in container => regular handling, update register
        if sounds[p] != []:
            val = sounds[p][0]
            del sounds[p][0]
            reg = cmd[1]
            registers[p][reg] = val

        # no sound => check other program
        else:

            # if other program is halting with empty container => deadlock
            if sounds[1 - p] == []:
                break
            else:
                p = 1 - p # it has sound => switch programs
                continue
        

    positions[p] += 1

c
```

# --- Day 19: A Series of Tubes ---

```python
puzzle = '''    |         
    |  +--+   
    A  |  C   
F---|----E|--+
    |  |  |  D
    +B-+  +--+'''
with open('19.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ list(line) for line in puzzle ]
puzzle

################################################################################
# PART 1

import numpy as np
puzzle = np.array(puzzle)
R, C = puzzle.shape
# print('R', R, 'C', C)

pos = np.array([ 0, np.where(puzzle[0] == '|')[0][0] ])
dir = np.array([1, 0])
# print('Start pos:', pos, '; dir:', dir)

curr = puzzle[ pos[0], pos[1] ]
path = ''

iter = 0
while True:
    # print('Pos:', pos, '; dir:', dir, '; current:', curr)

    # encountering a letter => add it to the path
    if curr not in ['|', '-', '+']:
        path += curr

    # reaching a cross section
    if curr == '+':

        # moved horizontally up this point => move vertically
        if dir[0] == 0:

            # check up and move up if possible
            if 0 <= pos[0] - 1:
                up = puzzle[ pos[0] - 1 , pos[1] ]
                if up not in ['-', ' ']:
                    dir = np.array([-1, 0])

            # same logic down
            if pos[0] + 1 < R:
                down = puzzle[ pos[0] + 1 , pos[1] ]
                if down not in ['-', ' ']:
                    dir = np.array([1, 0])
        
            
        # moved vertically => turn to horizontal moves
        else:
            
            # check left and turn left if possible
            if 0 <= pos[1] - 1:
                left = puzzle[ pos[0] , pos[1] - 1 ]
                if left not in ['|', ' ']: # left should be '-', '+' or a letter
                    dir = np.array([0, -1])

            # same logic on the right
            if pos[1] + 1 < C:
                right = puzzle[ pos[0] , pos[1] + 1 ]
                if right not in ['|', ' ']:
                    dir = np.array([0, 1])

    # step
    pos += dir

    if (pos[0] < 0 or R <= pos[0]) or (pos[1] < 0 or C <= pos[1]): # cannot move outside the map
        break
    else:
        curr = puzzle[ pos[0] , pos[1] ]
        if curr == ' ': # cannot stray from path to OBLIVION
            break

    iter += 1

path, iter + 1
```
