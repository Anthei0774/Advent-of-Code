```python
%run /Users/Tatu/Desktop/Advent-of-Code/utils.ipynb

hailstones = """19, 13, 30 @ -2,  1, -2
18, 19, 22 @ -1, -1, -2
20, 25, 34 @ -2, -2, -4
12, 31, 28 @ -1, -2, -1
20, 19, 15 @  1, -5, -3"""

test_area = [7, 27]

with open("inputs/24.txt", "r") as f:
    hailstones = f.read()
    test_area = [200000000000000, 400000000000000]

####################################################################################################
# PART 1

hailstones = hailstones.split('\n')
for i, hs in enumerate(hailstones):
    hs = hs.replace(",", "").replace("@ ", "").split(' ')
    hs = [coord for coord in hs if coord != ""]
    hs = list(map(float, hs))
    hs = Object2D(Vector2D(hs[0], hs[1]), Vector2D(hs[3], hs[4]))
    hailstones[i] = hs

n = len(hailstones)

c = 0

for i in range(n):
    A = hailstones[i]
    for j in range(i + 1, n):
        B = hailstones[j]

        # solve the for times t1, t2: A.p + t1 * A.v = B.p + t2 * B.v
        # written in matrix form: [A.v, -B.v] * [t1, t2] = B.p - A.p
        v = Matrix22(A.v, -B.v)
        p = B.p - A.p
        if det(v) == 0:
            print("Hailstones' paths are parallel; they never intersect.\n")
            continue
        T = v.inverse() * p
        t1, t2 = T.x, T.y

        # calculate valid solutions
        if t1 < 0 and t2 < 0:
            print("Hailstones' paths crossed in the past for both hailstones.\n")
            continue
        if t1 < 0:
            print("Hailstones' paths crossed in the past for hailstone A.\n")
            continue
        if t2 < 0:
            print("Hailstones' paths crossed in the past for hailstone B.\n")
            continue
        C = A.p + A.v * t1
        if (C >= test_area[0]) and (C <= test_area[1]):
            print(f"Hailstones' paths will cross inside the test area (at {C}\n")
            c += 1
        else:
            print(f"Hailstones' paths will cross outside the test area (at {C}\n")

c
```
