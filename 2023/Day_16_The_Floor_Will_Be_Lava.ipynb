```python
puzzle = r'''.|...\....
|.-.\.....
.....|-...
........|.
..........
.........\
..../.\\..
.-.-/..|..
.|....-|.\
..//.|....'''

with open('16.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [list(line) for line in puzzle]
puzzle

###############################################################################
# PART 1-2

R = len(puzzle)
C = len(puzzle[0])

# construct a dict of mirror and splitter
mirrors_splitters = {}
for r in range(R):
    for c in range(C):
        pos = (r, c)
        char = puzzle[r][c]
        if char != '.': mirrors_splitters[pos] = char
mirrors_splitters

# function to pass around light and return number of energized areas
def compute(lights):
    global mirrors_splitters, R, C

    while True:
        new_lights = []

        # process lights
        processing = [(l, d) for (l, d) in lights if not lights[(l, d)]]
        while processing != []:
            (l, d) = processing.pop(0)
            # print(l, d)

            lights[(l, d)] = True
            new_l = None
            new_d = None

            if l not in mirrors_splitters: # passing through empty position
                new_lights.append(((l[0] + d[0], l[1] + d[1]), d))
            else:
                if (d[0] == 0 and mirrors_splitters[l] == '-') or (d[1] == 0 and mirrors_splitters[l] == '|'): # passing through splitter
                    new_lights.append(((l[0] + d[0], l[1] + d[1]), d))
                else:

                    if d[0] != 0 and mirrors_splitters[l] == '-':
                        new_lights.append(((l[0], l[1] - 1), (0, -1))) # left
                        new_lights.append(((l[0], l[1] + 1), (0, 1))) # right

                    elif d[1] != 0 and mirrors_splitters[l] == '|':
                        new_lights.append(((l[0] - 1, l[1]), (-1, 0))) # up
                        new_lights.append(((l[0] + 1, l[1]), (1, 0))) # down

                    elif mirrors_splitters[l] == '/':
                        new_lights.append(((l[0] - d[1], l[1] - d[0]), (-d[1], -d[0])))

                    elif mirrors_splitters[l] == '\\':
                        new_lights.append(((l[0] + d[1], l[1] + d[0]), (d[1], d[0])))

                    else:
                        assert False
        # print('Possible new lights:', new_lights)

        # filter new lights
        new_lights = [(l, d) for (l, d) in new_lights if 0 <= l[0] and l[0] < R and 0 <= l[1] and l[1] < C] # not moving off from map
        new_lights = [(l, d) for (l, d) in new_lights if (l, d) not in lights] # must be a completely new light
        # print('New lights filtered:', new_lights, '\n')

        if new_lights == []: break

        for (l, d) in new_lights: lights[(l, d)] = False

    return len(set(l for (l, d) in lights))

# PART 1
compute({((0, 0), (0, 1)): False})

# PART 2 - brute force it
M = 0
for r in range(R):
    print('r:', r, 'M:', M)
    m = compute({((r, 0), (0, 1)): False})
    if M < m: M = m
    m = compute({((r, C - 1), (0, -1)): False})
    if M < m: M = m
for c in range(C):
    print('c:', c, 'M:', M)
    m = compute({((0, c), (1, 0)): False})
    if M < m: M = m
    m = compute({((R - 1, c), (-1, 0)): False})
    if M < m: M = m
M
```
