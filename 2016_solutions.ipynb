# --- Day 4: Security Through Obscurity ---

```python
puzzle = '''aaaaa-bbb-z-y-x-123[abxyz]
a-b-c-d-e-f-g-h-987[abcde]
not-a-real-room-404[oarel]
totally-real-room-200[decoy]'''

with open('4.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')

for i, line in enumerate(puzzle):
    # print(line)

    room = {'name': '', 'sectorid': 0, 'checksum': ''}

    name = '-'.join(line.split('-')[:-1])
    # print(name)
    room['name'] = name

    sectorid = line.split('-')[-1].split('[')[0]
    # print(sectorid)
    room['sectorid'] = int(sectorid)

    checksum = line.split('[')[-1][:-1]
    room['checksum'] = checksum

    puzzle[i] = room

################################################################################
# PART 1

presumed_checksums = []

for room in puzzle:
    letter_counts = []

    # count letters in room name, take top 5 most occuring
    for char in set(room['name']):
        if char == '-': continue

        letter_counts.append([room['name'].count(char), char])

    # good thing I searched for the double sorting in day 3
    # https://scripteverything.com/python-2d-list-sort-by-multiple-columns-code-examples-no-imports-one-liners
    # - sign for descending sort in the counts, but ascending sort in letters
    letter_counts = sorted(letter_counts, key = lambda count: (-count[0], count[1]))
    letter_counts = letter_counts[:5]
    presumed_checksums.append(letter_counts)

# concatenate the 5 letters into a single word
presumed_checksums = [ ''.join([char[1] for char in line]) for line in presumed_checksums ]


s = 0
for i, p_checksum in enumerate(presumed_checksums):
    room = puzzle[i]
    room['decoy'] = True

    # room is not a decoy
    if p_checksum == room['checksum']:
        room['decoy'] = False
        s += room['sectorid']

s

################################################################################
# PART 2

abc = 'abcdefghijklmnopqrstuvwxyz'

for room in puzzle:
    res = ''

    if not room['decoy']:
        # print(room)

        for char in room['name']:
            if char == '-':
                res += ' '
                continue 
            i = abc.index(char)
            i = (i + room['sectorid'] % 26) % 26
            res += abc[i]

        # print(res)
        if 'northpole' in res:
            print(room, res)
```

```python
# part 1 example
puzzle = '''ADVENT
A(1x5)BC
(3x3)XYZ
A(2x2)BCD(2x2)EFG
(6x1)(1x3)A
X(8x2)(3x3)ABCY'''

# part 2 example
puzzle = '''(3x3)XYZ
X(8x2)(3x3)ABCY
(27x12)(20x12)(13x14)(7x10)(1x12)A
(25x3)(3x3)ABC(2x3)XY(5x2)PQRSTX(18x9)(3x2)TWO(5x7)SEVEN'''

with open('9.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle

################################################################################
# PART 1

def decompress_line_part1(line):
    # print('Line:', line)

    # find first parentheses
    start = line.find('(')
    if start != -1:
        end = line.find(')', start)

        prefix = line[: start]
        # print('Prefix:', prefix)

        # extract sequence length and repetition times from the parenthesis
        [length, repeat] = list(map(int, line[start + 1 : end].split('x')))


        suffix = line[end + 1 + length :]
        suffix = decompress_line_part1(suffix)
        # print('Suffix:', suffix)

        repeated = line[end + 1 : end + 1 + length] * repeat
        # print('Repeated:', repeated)

        new_line = prefix + repeated + suffix

        return new_line

    # no parentheses = nothing to decompress
    else:
        # print('No decompression')
        return line


decompressed = []

for line in puzzle:
    new_line = decompress_line_part1(line)
    # print('Decompressed line:', new_line)
    # print(len(new_line))
    # print('\n---\n')


################################################################################
# PART 2

def decompress_line_part2(line):
    # print('Line:', line)

    # find first parentheses
    start = line.find('(')
    if start != -1:
        end = line.find(')', start)

        prefix = line[: start]
        l_prefix = len(prefix)
        # print('Prefix:', prefix, '; len:', l_prefix)
        

        # extract sequence length and repetition times from the parenthesis
        [length, repeat] = list(map(int, line[start + 1 : end].split('x')))


        # need to decompress both suffix and region after parenthesis
        # return only the length of the decompressed line

        # 1st suffix
        suffix = line[end + 1 + length :]
        l_suffix = decompress_line_part2(suffix)
        # print('Suffix:', suffix, '; len:', l_suffix)

        # 2nd inside
        inside = line[end + 1 : end + 1 + length]
        l_inside = decompress_line_part2(inside)
        l_inside_repeated = l_inside * repeat
        # print('Inside:', inside, '; len:', l_inside, '; repeated len:', l_inside_repeated)

        # sum up
        l_total = l_prefix + l_inside_repeated + l_suffix
        return l_total

    # no parentheses = nothing to decompress
    else:
        # print('No decompression')
        return len(line)


for line in puzzle:
    l_total = decompress_line_part2(line)
    print('Total decompression len:', l_total)
    # print('\n---\n')
```

# --- Day 10: Balance Bots ---

```python
puzzle = '''value 5 goes to bot 2
bot 2 gives low to bot 1 and high to bot 0
value 3 goes to bot 1
bot 1 gives low to output 1 and high to bot 0
bot 0 gives low to output 2 and high to output 0
value 2 goes to bot 2'''

with open('10.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')

# sort first by bot / value, then by index
puzzle = sorted(puzzle, key = lambda x: (x.split(' ')[0], int(x.split(' ')[1])) )
puzzle

bots = []

for line in puzzle:
    # print(line)

    line_type = line.split(' ')[0]
    if line_type == 'bot':
        bot = {
            'low': int(line.split(' ')[6])
            , 'low_bot': line.split(' ')[5] == 'bot'
            , 'high': int(line.split(' ')[-1])
            , 'high_bot': line.split(' ')[-2] == 'bot'
            , 'values': []
            }

        # print(bot)
        bots.append(bot)

    else:
        value = int(line.split(' ')[1])
        index = int(line.split(' ')[-1])

        bots[index]['values'].append(value)
        bots[index]['values'].sort()

bots

################################################################################
# PART 1

outputs = {}

found_bot = False

# iterate until the bot found with [17, 61] ready to be given to other bots / outputs
iter = 0
while not found_bot:
    # print('Iteration:', iter)
    iter += 1

    givers = [ (i, b) for i, b in enumerate(bots) if len(b['values']) == 2 ]

    for (i, b) in givers:
        # print(b)

        if b['values'] == [17, 61]:
            found_bot = True
            break

        # if low target is bot => pass values, then sort
        if b['low_bot']:
            bots[ b['low'] ]['values'].append(b['values'][0])
            bots[ b['low'] ]['values'].sort()
        # low target is output => initiate new output channel or append to the end
        else:
            if b['low'] not in outputs:
                outputs[b['low']] = [b['values'][0]]
            else:
                outputs[b['low']].append(b['values'][0])

        # same logic on high end
        if b['high_bot']:
            bots[ b['high'] ]['values'].append(b['values'][1])
            bots[ b['high'] ]['values'].sort()
        else:
            if b['high'] not in outputs:
                outputs[b['high']] = [b['values'][1]]
            else:
                outputs[b['high']].append(b['values'][1])

        # empty bot values
        b['values'] = []

i

################################################################################
# PART 2

# continue iterating but with a different stopping critera (= no more moves)
iter = 0
while True:
    # print('Iteration:', iter)
    iter += 1

    givers = [(i, b) for i, b in enumerate(bots) if len(b['values']) == 2]
    if givers == []: break # stable state of bots, no values passed

    for (i, b) in givers:
        # print(b)

        # if low target is bot => pass values, then sort
        if b['low_bot']:
            bots[ b['low'] ]['values'].append(b['values'][0])
            bots[ b['low'] ]['values'].sort()
        # low target is output => initiate new output channel or append to the end
        else:
            if b['low'] not in outputs:
                outputs[b['low']] = [b['values'][0]]
            else:
                outputs[b['low']].append(b['values'][0])

        # same logic on high end
        if b['high_bot']:
            bots[ b['high'] ]['values'].append(b['values'][1])
            bots[ b['high'] ]['values'].sort()
        else:
            if b['high'] not in outputs:
                outputs[b['high']] = [b['values'][1]]
            else:
                outputs[b['high']].append(b['values'][1])

        # empty bot values
        b['values'] = []


outputs[0][0] * outputs[1][0] * outputs[2][0]
```

```python
'''The first floor contains a hydrogen-compatible microchip and a lithium-compatible microchip.
The second floor contains a hydrogen generator.
The third floor contains a lithium generator.
The fourth floor contains nothing relevant.'''

'''The first floor contains a thulium generator, a thulium-compatible microchip, a plutonium generator, and a strontium generator.
The second floor contains a plutonium-compatible microchip and a strontium-compatible microchip.
The third floor contains a promethium generator, a promethium-compatible microchip, a ruthenium generator, and a ruthenium-compatible microchip.
The fourth floor contains nothing relevant.'''

# encoding the puzzle input by hand
s = '0,HydrM-LithM,HydrG,LithG,'
# s = '0,ThulG-ThulM-PlutG-StroG,PlutM-StroM,PromG-PromM-RuthG-RuthM'

states = {}
states[s] = {'neighbours': [], 'processed': False}

################################################################################
# PART 1

iter = 0
while True:
    print('Iteration:', iter)

    not_processed_states = [ s for s in states if not states[s]['processed'] ]
    if not_processed_states == []: break

    for nps in not_processed_states:
        states[nps]['processed'] = True

        # print('A non-processed state:', nps)

        level_items = nps.split(',')[1 :]
        level_items = [ level.split('-') if level != '' else [] for level in level_items ]
        # print('Levels with items:', level_items)

        elevator_level = int(nps.split(',')[0])
        # print('Elevator level:', elevator_level)

        elevator_level_items = level_items[elevator_level]
        # print('Elevator level items:', elevator_level_items)

        # the elevator will bring up 1 or 2 items
        item_combinations = []
        for i in range(len(elevator_level_items)):
            item_combinations.append([elevator_level_items[i]])

            for j in range(i + 1, len(elevator_level_items)):
                item_combinations.append([elevator_level_items[i], elevator_level_items[j]])
        # print('Item combinations taken up / down:', item_combinations, '\n')

        # decide directions
        directions = []
        if elevator_level != 3: directions.append(1)
        if elevator_level != 0: directions.append(-1)

        # construct new states from directions and item combinations
        new_states = {}

        for d in directions:
            for ic in item_combinations:
                # print('Combination direction + items:', 'up' if d == 1 else 'down', ',', ic)

                # start constructing new state by adding elevator level
                elevator_target_level = elevator_level + d
                ns = str(elevator_target_level)
                
                # iterate through level
                for i in range(4):

                    # remove items from elevator level
                    if i == elevator_level:
                        remaining_items_on_level = [ item for item in level_items[i] if item not in ic ]
                        # print('Remaining items on level:', remaining_items_on_level)
                        
                        ns += (',' + '-'.join(remaining_items_on_level))
                        continue

                    # add items to elevator target level
                    if i == elevator_target_level:
                        new_items_on_level = level_items[i] + ic
                        new_items_on_level = sorted(new_items_on_level)
                        # print('New items on level:', new_items_on_level)

                        ns += (',' + '-'.join(new_items_on_level))
                        continue

                    # copy the level otherwise
                    ns += (',' + '-'.join(level_items[i]))
                
                # state constructed
                new_states[ns] = {'neighbours': [nps], 'processed': False}
                # print(ns, '\n')

        # print('--- Checking new states ---')

        # checking if the new states are valid
        for ns in new_states:
            # print(ns)

            valid = True
            level_items = ns.split(',')[1 :]
            level_items = [ level.split('-') if level != '' else [] for level in level_items ]
            # print(level_items)

            # iterating through levels of a state
            for level in level_items:
                # print(level)

                # checking all possible combinations of items on level
                for i in range(len(level)):
                    for j in range(i + 1, len(level)):
                        item_i = level[i]
                        item_j = level[j]
                        
                        # reject state if microchip is present without corresponding generator but with other generator
                        if item_i[-1] == 'G' and item_j[-1] == 'M' and item_i[: -1] != item_j[: -1]:
                            valid = False
                        if item_i[-1] == 'M' and item_j[-1] == 'G' and item_i[: -1] != item_j[: -1]:
                            valid = False

            # remove false state
            if not valid:
                new_states[ns] = False

        # shrink new states to only valid ones
        new_states = {ns : new_states[ns] for ns in new_states if new_states[ns] != False }

        # insert new states
        for ns in new_states:

            # real new state
            if ns not in states:
                states[ns] = new_states[ns]
                for nb in new_states[ns]['neighbours']:
                    states[nb]['neighbours'].append(ns)
                    # assert len(states[nb]['neighbours']) == len(set(states[nb]['neighbours']))

            # already existing (= backstep happened) ??? what to do here
            else:
                pass
                



    if iter == 100:
        break
    iter += 1
```

```python
def poly(x, y): return x*x + 3*x + 2*x*y + y + y*y

puzzle = 10
size = (7, 10)

puzzle = 1362 # puzzle input
size = (50, 50) # setting it big enough for part 1
# size = (100, 100) # setting it bigger for part 2

grid = [ [ poly(x, y) + puzzle for x in range(size[1]) ] for y in range(size[0]) ] # generate grid with poly and office designer's favourite number
grid = [ [ format(num, 'b').count('1') for num in line ] for line in grid ] # into binary, then count ones
grid = [ [ int(num % 2 == 1) for num in line ] for line in grid ]
grid


################################################################################
# PART 1 - 2

reached_points = {
    (1, 1) : {'checked': False, 'cost' : 0}
}

# main idea to implement a breadth first search - finding shortest path to any target

# fail switch if while-loop goes out of hand
iter = 0
while True:
    
    all_new_points = []

    # iterate through reached points, which are not checked for neighbours
    for coords in reached_points:
        rp = reached_points[coords]
        if not rp['checked']:
            rp['checked'] = True

            # find valid neighbours
            new_points = [(coords[0] + 1, coords[1]), (coords[0] - 1, coords[1]), (coords[0], coords[1] + 1), (coords[0], coords[1] - 1)] # step left-right-up-down
            new_points = [ np for np in new_points if np[0] >= 0 and np[0] < size[1] and np[1] >= 0 and np[1] < size[0] and grid[np[1]][np[0]] == 0 ] # staying on the grid and not wall
            # print(new_points)

            # save parent and list of points reached from there
            all_new_points.append({ 'parent': rp, 'new_points': new_points })

    
    # no new points = maze explored
    if all_new_points == []:
        break
    # print(all_new_points)


    # for each new point, append to reached (if not already in) with cost increased
    for a_np in all_new_points:
        for np in a_np['new_points']:
            if np not in reached_points:
                reached_points[np] = {'checked': False, 'cost' : rp['cost'] + 1}


    # failsafe break
    if iter == 1000:
        break
    iter += 1


reached_points[ (7, 4) ] # example answer
reached_points[ (31, 39) ] # part 1 answer
len({ coords : reached_points[coords] for coords in reached_points if reached_points[coords]['cost'] <= 50 }) # part 2 answer
```
