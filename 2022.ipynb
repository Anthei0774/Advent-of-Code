# --- Day 1: Calorie Counting ---

```python
puzzle = '''1000
2000
3000

4000

5000
6000

7000
8000
9000

10000'''
with open('1.txt') as f: puzzle = f.read()


puzzle = puzzle.split('\n\n')
puzzle = [ elf.split('\n') for elf in puzzle ]
puzzle = [ list(map(int, elf)) for elf in puzzle ]
puzzle

################################################################################
# PART 1

max( sum(elf) for elf in puzzle )

################################################################################
# PART 2

top3 = sorted( sum(elf) for elf in puzzle )
sum(top3[-3:])
```

# --- Day 2: Rock Paper Scissors ---

```python
puzzle = '''A Y
B X
C Z'''
with open('2.txt') as f: puzzle = f.read()

puzzle = puzzle.replace('A', '1').replace('X', '1')
puzzle = puzzle.replace('B', '2').replace('Y', '2')
puzzle = puzzle.replace('C', '3').replace('Z', '3')

puzzle = puzzle.split('\n')
puzzle = [ round.split(' ') for round in puzzle ]
puzzle = [ tuple(map(int, round)) for round in puzzle ]
puzzle

################################################################################
# PART 1

score = 0
for round in puzzle:
    
    score += round[1] # add shape score

    if (round[0] % 3) + 1 == round[1]: # cases of 1-2, 2-3, 3-1; winning
        score += 6
    elif round[0] == round[1]: # draw
        score += 3
    else:
        continue

score

################################################################################
# PART 2

# shape and win score calculation switches places
score = 0
for round in puzzle:

    score += ((round[1] - 1) * 3) # transform 1-2-3 into 0-3-6, get win score

    if round[1] == 1: # need to lose, that is 1-3; 2-1; 3-2
        s = round[0] - 1
        if s == 0: s = 3
        score += s
    elif round[1] == 2: # draw
        score += round[0]
    else:
        score += ((round[0] % 3) + 1)

score
```
