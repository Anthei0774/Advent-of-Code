# --- Day 2: Rock Paper Scissors ---

```python
puzzle = '''A Y
B X
C Z'''
with open('2.txt') as f: puzzle = f.read()

puzzle = puzzle.replace('A', '1').replace('X', '1')
puzzle = puzzle.replace('B', '2').replace('Y', '2')
puzzle = puzzle.replace('C', '3').replace('Z', '3')

puzzle = puzzle.split('\n')
puzzle = [ round.split(' ') for round in puzzle ]
puzzle = [ tuple(map(int, round)) for round in puzzle ]
puzzle

################################################################################
# PART 1

score = 0
for round in puzzle:
    
    score += round[1] # add shape score

    if (round[0] % 3) + 1 == round[1]: # cases of 1-2, 2-3, 3-1; winning
        score += 6
    elif round[0] == round[1]: # draw
        score += 3
    else:
        continue

score

################################################################################
# PART 2

# shape and win score calculation switches places
score = 0
for round in puzzle:

    score += ((round[1] - 1) * 3) # transform 1-2-3 into 0-3-6, get win score

    if round[1] == 1: # need to lose, that is 1-3; 2-1; 3-2
        s = round[0] - 1
        if s == 0: s = 3
        score += s
    elif round[1] == 2: # draw
        score += round[0]
    else:
        score += ((round[0] % 3) + 1)

score
```

# --- Day 3: Rucksack Reorganization ---

```python
puzzle = '''vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw'''
with open('3.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, rucksack in enumerate(puzzle):
    l = len(rucksack)
    assert l % 2 == 0
    puzzle[i] = [ rucksack[: l // 2], rucksack[l // 2 :] ]
puzzle

################################################################################
# PART 1

abc = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'


#
prio = 0
for rucksack in puzzle:
    comp0, comp1 = rucksack
    
    i = set(comp0).intersection(set(comp1))
    assert len(i) == 1
    i = list(i)[0]

    prio += (abc.index(i) + 1)
prio

################################################################################
# PART 2

prio = 0
for i in range(0, len(puzzle), 3):

    group = puzzle[i : i + 3]
    group = [ ''.join(rucksack) for rucksack in group ]
    
    m = set(group[0])
    m = m.intersection(set(group[1]))
    m = m.intersection(set(group[2]))
    assert len(m) == 1
    
    m = list(m)[0]
    prio += (abc.index(m) + 1)
prio
```

# --- Day 5: Supply Stacks ---

```python
puzzle = '''    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2'''
with open('5.txt') as f: puzzle = f.read()


puzzle = puzzle.split('\n\n')

def process_puzzle(puzzle): # processing stacks of crates into dict format --- as function, so its reusable = can reset stacks
    stacks = puzzle[0].split('\n')
    stacks = [ list(level) for level in stacks ] # to 2D char array
    stacks = list(zip(*stacks)) # transpose, but array of tuples, https://stackoverflow.com/questions/6473679/transpose-list-of-lists
    stacks = list(map(list, stacks))
    stacks = [ crates for crates in stacks if crates[-1] != ' ' ]
    stacks = [ [ c for c in crates if c != ' ' ] for crates in stacks ]
    stacks = { int(crates[-1]) : crates[-2::-1] for crates in stacks } # https://stackoverflow.com/questions/52657192/how-to-loop-a-list-in-reverse-order-and-skip-last-element
    return stacks

stacks = process_puzzle(puzzle)

# processing instructions into list of dicts format
instructions = puzzle[1].split('\n')
instructions = [ i.split(' ') for i in instructions ]
instructions = [ { 'source' : int(i[3]), 'destination' : int(i[-1]), 'count' : int(i[1]) } for i in instructions ]
instructions

################################################################################
# PART 1

# print(stacks, '\n')
for i in instructions:
    # print(i)
    
    crates = stacks[ i['source'] ]

    tmp = []
    for _ in range(i['count']):
        c = crates.pop(-1)
        tmp.append(c)
    
    stacks[ i['source'] ] = crates
    stacks[ i['destination'] ] += tmp
    # print(stacks, '\n')

crates = list(stacks.values())
crates = [ c[-1] for c in crates ]
crates = ''.join(crates)
crates

################################################################################
# PART 2

# reset
stacks = process_puzzle(puzzle)
stacks

# same operation, but without pop-append reversing
# print(stacks, '\n')
for i in instructions:
    # print(i)
    
    crates = stacks[ i['source'] ]

    tmp = crates[-i['count']:]
    crates = crates[:-i['count']]
    
    stacks[ i['source'] ] = crates
    stacks[ i['destination'] ] += tmp
    # print(stacks, '\n')

crates = list(stacks.values())
crates = [ c[-1] for c in crates ]
crates = ''.join(crates)
crates
```

# --- Day 7: No Space Left On Device ---

```python
puzzle = '''$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k'''
with open('7.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle = puzzle[1:]
puzzle

################################################################################
# PART 1

# organize files in tree structure, need list comprehension as input files contains files and directories with the same name
tree = [ { 'name' : 'root', 'prev' : None, 'type' : 'dir', 'children' : [] } ]
current = 0

for line in puzzle:

    # check if command or item
    if line[0] == '$':

        # if command then ls or cd ==> only handle cd
        if line[1] == 'ls':
            continue
        else:

            if line[2] == '..':
                current = tree[current]['prev'] # stepping back is easy
            else:

                # execute small search in indices, with finding matching name
                for c in tree[current]['children']: 
                    if tree[c]['name'] == line[2]:
                        current = c
                        break

    else:
        # if item, then make a general one
        item = { 'name' : line[1], 'prev' : current, 'type' : line[0] }

        # check if subdir or file, add property
        if item['type'] == 'dir':
            item['children'] = []
        else:
            item['size'] = int(line[0])

        # add to tree
        tree[current]['children'].append( len(tree) )
        tree.append(item)

tree


# recursize function to calculate dir sizes
def calculate_size(idx):
    global tree

    if tree[idx]['type'] == 'dir': # calculate size of children, when dir
        s = 0
        for c in tree[idx]['children']: s += calculate_size(c) # indices allow easy recursive call
        tree[idx]['size'] = s
        return s

    else: # return immediate size, when file
        return tree[idx]['size']

# run
calculate_size(0)
sum(node['size'] for node in tree if node['type'] == 'dir' and node['size'] <= 100000) # return dirs with small enough size

################################################################################
# PART 2

# messing around for 30+ mins in PART 1, then solving PART 2 in 2 mins... :)

unused = 70000000 - tree[0]['size'] # total - root/used size
min( node['size'] for node in tree if node['type'] == 'dir' and node['size'] + unused >= 30000000 ) # search for dirs, that - upon deletion - would free enough memory size for needed, choose minimum
```

# --- Day 8: Treetop Tree House ---

```python
puzzle = '''30373
25512
65332
33549
35390'''
with open('8.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = list(map(list, puzzle))

import numpy as np
puzzle = np.array(puzzle, int)
puzzle

################################################################################
# PART 1

size = puzzle.shape

v = 2 * (sum(size) - 2)
for i in range(1, size[0] - 1):
    for j in range(1, size[1] - 1):
        t = puzzle[i, j]
        v += all(puzzle[:i, j] < t) or all(puzzle[i + 1:, j] < t) or all(puzzle[i, :j] < t) or all(puzzle[i, j + 1:] < t)
v

################################################################################
# PART 2

S = 0
for i in range(1, size[0] - 1): # the scenic score of a tree on sides would be always 0, because of multiplying values
    for j in range(1, size[1] - 1): # enough to select from not-side-trees

        # tree height under consideration
        t = puzzle[i, j]
        # print(i, j, t)

        # logic in each direction:
        # 1. create array of trees going from current to distant (need to reverse in some cases)
        # 2. find the first tree which is as tall as the current
        # 3. calculate distance and multiply
        s = 1

        # up
        u = puzzle[:i, j][::-1] # need to revese, so it goes bottom-up
        m = 1
        if len(u) >= 2:
            while m < len(u) and u[m - 1] < t:
                m += 1
        # print('u', m)
        s *= m

        # down
        d = puzzle[i+1:, j]
        m = 1
        if len(d) >= 2:
            while m < len(d) and d[m - 1] < t:
                m += 1
        # print('d', m)
        s *= m

        # left
        l = puzzle[i, :j][::-1] # need to reverse, so it goes right-to-left
        m = 1
        if len(l) >= 2:
            while m < len(l) and l[m - 1] < t:
                m += 1
        # print('l', m)
        s *= m

        # right
        r = puzzle[i, j+1:]
        m = 1
        if len(r) >= 2:
            while m < len(r) and r[m - 1] < t:
                m += 1
        # print('r', m)
        s *= m

        # maximum search for the highest product
        # print(s)
        if S < s: S = s

S
```

# --- Day 9: Rope Bridge ---

```python
puzzle = '''R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2''' # part 1 = 13, part 2 = 2
puzzle = '''R 5
U 8
L 8
D 3
R 17
D 10
L 25
U 20''' # part 2 = 36
with open('9.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ step.split(' ') for step in puzzle ]
puzzle = [ [ step[0], int(step[1]) ] for step in puzzle  ]
puzzle

################################################################################
# PART 1

H = (0, 0)
T = (0, 0)

visited = [ (0, 0) ]

for step in puzzle:

    dir = None
    if step[0] == 'L': dir = (-1, 0)
    if step[0] == 'R': dir = (1, 0)
    if step[0] == 'U': dir = (0, 1)
    if step[0] == 'D': dir = (0, -1)

    for i in range(step[1]):
        H = (H[0] + dir[0], H[1] + dir[1])
        
        # the tail must follow the head
        # diff shows the H's relative position to T
        diff = (H[0] - T[0], H[1] - T[1])

        if diff == (0, 2):
            T = (T[0], T[1] + 1)
        if diff == (0, -2):
            T = (T[0], T[1] - 1)
        if diff == (2, 0):
            T = (T[0] + 1, T[1])
        if diff == (-2, 0):
            T = (T[0] - 1, T[1])
        if diff == (2, 1) or diff == (1, 2):
            T = (T[0] + 1, T[1] + 1)
        if diff == (-2, 1) or diff == (-1, 2):
            T = (T[0] - 1, T[1] + 1)
        if diff == (-2, -1) or diff == (-1, -2):
            T = (T[0] - 1, T[1] - 1)
        if diff == (2, -1) or diff == (1, -2):
            T = (T[0] + 1, T[1] - 1)

        # print('H:', H, 'T:', T)

        if T not in visited: visited.append(T)

len(visited)

################################################################################
# PART 2

knots = [ (0, 0) for i in range(10) ]
knots

visited = [ (0, 0) ]

for step in puzzle:

    dir = None
    if step[0] == 'L': dir = (-1, 0)
    if step[0] == 'R': dir = (1, 0)
    if step[0] == 'U': dir = (0, 1)
    if step[0] == 'D': dir = (0, -1)

    for i in range(step[1]):
        # print('i:', i)
        H = knots[0]
        H = (H[0] + dir[0], H[1] + dir[1])
        knots[0] = H
        
        # the tail must follow the head
        # diff shows the knot[i - 1]'s relative position to knot[i]
        for j in range(1, 10):
            # print('j:', j)
            H = knots[j - 1]
            T = knots[j]
            diff = (H[0] - T[0], H[1] - T[1])

            if diff == (0, 0):
                break
            if diff == (0, 2):
                T = (T[0], T[1] + 1)
            if diff == (0, -2):
                T = (T[0], T[1] - 1)
            if diff == (2, 0):
                T = (T[0] + 1, T[1])
            if diff == (-2, 0):
                T = (T[0] - 1, T[1])
            # had to add 2-diag-diff to the last 4 cases
            if diff == (2, 1) or diff == (1, 2) or diff == (2, 2):
                T = (T[0] + 1, T[1] + 1)
            if diff == (-2, 1) or diff == (-1, 2) or diff == (-2, 2):
                T = (T[0] - 1, T[1] + 1)
            if diff == (-2, -1) or diff == (-1, -2) or diff == (-2, -2):
                T = (T[0] - 1, T[1] - 1)
            if diff == (2, -1) or diff == (1, -2) or diff == (2, -2):
                T = (T[0] + 1, T[1] - 1)

            knots[j] = T

        # print(knots)

        if knots[-1] not in visited: visited.append(knots[-1])

len(visited)
```
