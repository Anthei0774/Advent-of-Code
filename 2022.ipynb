# --- Day 1: Calorie Counting ---

```python
puzzle = '''1000
2000
3000

4000

5000
6000

7000
8000
9000

10000'''
with open('1.txt') as f: puzzle = f.read()


puzzle = puzzle.split('\n\n')
puzzle = [ elf.split('\n') for elf in puzzle ]
puzzle = [ list(map(int, elf)) for elf in puzzle ]
puzzle

################################################################################
# PART 1

max( sum(elf) for elf in puzzle )

################################################################################
# PART 2

top3 = sorted( sum(elf) for elf in puzzle )
sum(top3[-3:])
```

# --- Day 2: Rock Paper Scissors ---

```python
puzzle = '''A Y
B X
C Z'''
with open('2.txt') as f: puzzle = f.read()

puzzle = puzzle.replace('A', '1').replace('X', '1')
puzzle = puzzle.replace('B', '2').replace('Y', '2')
puzzle = puzzle.replace('C', '3').replace('Z', '3')

puzzle = puzzle.split('\n')
puzzle = [ round.split(' ') for round in puzzle ]
puzzle = [ tuple(map(int, round)) for round in puzzle ]
puzzle

################################################################################
# PART 1

score = 0
for round in puzzle:
    
    score += round[1] # add shape score

    if (round[0] % 3) + 1 == round[1]: # cases of 1-2, 2-3, 3-1; winning
        score += 6
    elif round[0] == round[1]: # draw
        score += 3
    else:
        continue

score

################################################################################
# PART 2

# shape and win score calculation switches places
score = 0
for round in puzzle:

    score += ((round[1] - 1) * 3) # transform 1-2-3 into 0-3-6, get win score

    if round[1] == 1: # need to lose, that is 1-3; 2-1; 3-2
        s = round[0] - 1
        if s == 0: s = 3
        score += s
    elif round[1] == 2: # draw
        score += round[0]
    else:
        score += ((round[0] % 3) + 1)

score
```

# --- Day 3: Rucksack Reorganization ---

```python
puzzle = '''vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw'''
with open('3.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, rucksack in enumerate(puzzle):
    l = len(rucksack)
    assert l % 2 == 0
    puzzle[i] = [ rucksack[: l // 2], rucksack[l // 2 :] ]
puzzle

################################################################################
# PART 1

abc = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'


#
prio = 0
for rucksack in puzzle:
    comp0, comp1 = rucksack
    
    i = set(comp0).intersection(set(comp1))
    assert len(i) == 1
    i = list(i)[0]

    prio += (abc.index(i) + 1)
prio

################################################################################
# PART 2

prio = 0
for i in range(0, len(puzzle), 3):

    group = puzzle[i : i + 3]
    group = [ ''.join(rucksack) for rucksack in group ]
    
    m = set(group[0])
    m = m.intersection(set(group[1]))
    m = m.intersection(set(group[2]))
    assert len(m) == 1
    
    m = list(m)[0]
    prio += (abc.index(m) + 1)
prio
```

# --- Day 4: Camp Cleanup ---

```python
puzzle = '''2-4,6-8
2-3,4-5
5-7,7-9
2-8,3-7
6-6,4-6
2-6,4-8'''
with open('4.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ pair.replace(',', '-').split('-') for pair in puzzle ]
puzzle = [ tuple(map(int, pair)) for pair in puzzle ]
puzzle

################################################################################
# PART 1

c = 0
for pair in puzzle:
    # fully contains if indicies are like 0-2-3-1 or 2-0-1-3
    if (pair[0] <= pair[2] and pair[3] <= pair[1]) or (pair[2] <= pair[0] and pair[1] <= pair[3]):
        c += 1
c

################################################################################
# PART 2

c = 0
for pair in puzzle:
    # partially contains if indices are like 0-2-1 or 0-3-1 or 2-0-3 or 2-1-3
    if (pair[0] <= pair[2] and pair[2] <= pair[1]) or (pair[0] <= pair[3] and pair[3] <= pair[1]) or (pair[2] <= pair[0] and pair[0] <= pair[3]) or (pair[2] <= pair[1] and pair[1] <= pair[3]):
        c += 1
c
```
