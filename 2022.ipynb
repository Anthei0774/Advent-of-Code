# --- Day 2: Rock Paper Scissors ---

```python
puzzle = '''A Y
B X
C Z'''
with open('2.txt') as f: puzzle = f.read()

puzzle = puzzle.replace('A', '1').replace('X', '1')
puzzle = puzzle.replace('B', '2').replace('Y', '2')
puzzle = puzzle.replace('C', '3').replace('Z', '3')

puzzle = puzzle.split('\n')
puzzle = [ round.split(' ') for round in puzzle ]
puzzle = [ tuple(map(int, round)) for round in puzzle ]
puzzle

################################################################################
# PART 1

score = 0
for round in puzzle:
    
    score += round[1] # add shape score

    if (round[0] % 3) + 1 == round[1]: # cases of 1-2, 2-3, 3-1; winning
        score += 6
    elif round[0] == round[1]: # draw
        score += 3
    else:
        continue

score

################################################################################
# PART 2

# shape and win score calculation switches places
score = 0
for round in puzzle:

    score += ((round[1] - 1) * 3) # transform 1-2-3 into 0-3-6, get win score

    if round[1] == 1: # need to lose, that is 1-3; 2-1; 3-2
        s = round[0] - 1
        if s == 0: s = 3
        score += s
    elif round[1] == 2: # draw
        score += round[0]
    else:
        score += ((round[0] % 3) + 1)

score
```

# --- Day 5: Supply Stacks ---

```python
puzzle = '''    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2'''
with open('5.txt') as f: puzzle = f.read()


puzzle = puzzle.split('\n\n')

def process_puzzle(puzzle): # processing stacks of crates into dict format --- as function, so its reusable = can reset stacks
    stacks = puzzle[0].split('\n')
    stacks = [ list(level) for level in stacks ] # to 2D char array
    stacks = list(zip(*stacks)) # transpose, but array of tuples, https://stackoverflow.com/questions/6473679/transpose-list-of-lists
    stacks = list(map(list, stacks))
    stacks = [ crates for crates in stacks if crates[-1] != ' ' ]
    stacks = [ [ c for c in crates if c != ' ' ] for crates in stacks ]
    stacks = { int(crates[-1]) : crates[-2::-1] for crates in stacks } # https://stackoverflow.com/questions/52657192/how-to-loop-a-list-in-reverse-order-and-skip-last-element
    return stacks

stacks = process_puzzle(puzzle)

# processing instructions into list of dicts format
instructions = puzzle[1].split('\n')
instructions = [ i.split(' ') for i in instructions ]
instructions = [ { 'source' : int(i[3]), 'destination' : int(i[-1]), 'count' : int(i[1]) } for i in instructions ]
instructions

################################################################################
# PART 1

# print(stacks, '\n')
for i in instructions:
    # print(i)
    
    crates = stacks[ i['source'] ]

    tmp = []
    for _ in range(i['count']):
        c = crates.pop(-1)
        tmp.append(c)
    
    stacks[ i['source'] ] = crates
    stacks[ i['destination'] ] += tmp
    # print(stacks, '\n')

crates = list(stacks.values())
crates = [ c[-1] for c in crates ]
crates = ''.join(crates)
crates

################################################################################
# PART 2

# reset
stacks = process_puzzle(puzzle)
stacks

# same operation, but without pop-append reversing
# print(stacks, '\n')
for i in instructions:
    # print(i)
    
    crates = stacks[ i['source'] ]

    tmp = crates[-i['count']:]
    crates = crates[:-i['count']]
    
    stacks[ i['source'] ] = crates
    stacks[ i['destination'] ] += tmp
    # print(stacks, '\n')

crates = list(stacks.values())
crates = [ c[-1] for c in crates ]
crates = ''.join(crates)
crates
```

# --- Day 7: No Space Left On Device ---

```python
puzzle = '''$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k'''
with open('7.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle = puzzle[1:]
puzzle

################################################################################
# PART 1

# organize files in tree structure, need list comprehension as input files contains files and directories with the same name
tree = [ { 'name' : 'root', 'prev' : None, 'type' : 'dir', 'children' : [] } ]
current = 0

for line in puzzle:

    # check if command or item
    if line[0] == '$':

        # if command then ls or cd ==> only handle cd
        if line[1] == 'ls':
            continue
        else:

            if line[2] == '..':
                current = tree[current]['prev'] # stepping back is easy
            else:

                # execute small search in indices, with finding matching name
                for c in tree[current]['children']: 
                    if tree[c]['name'] == line[2]:
                        current = c
                        break

    else:
        # if item, then make a general one
        item = { 'name' : line[1], 'prev' : current, 'type' : line[0] }

        # check if subdir or file, add property
        if item['type'] == 'dir':
            item['children'] = []
        else:
            item['size'] = int(line[0])

        # add to tree
        tree[current]['children'].append( len(tree) )
        tree.append(item)

tree


# recursize function to calculate dir sizes
def calculate_size(idx):
    global tree

    if tree[idx]['type'] == 'dir': # calculate size of children, when dir
        s = 0
        for c in tree[idx]['children']: s += calculate_size(c) # indices allow easy recursive call
        tree[idx]['size'] = s
        return s

    else: # return immediate size, when file
        return tree[idx]['size']

# run
calculate_size(0)
sum(node['size'] for node in tree if node['type'] == 'dir' and node['size'] <= 100000) # return dirs with small enough size

################################################################################
# PART 2

# messing around for 30+ mins in PART 1, then solving PART 2 in 2 mins... :)

unused = 70000000 - tree[0]['size'] # total - root/used size
min( node['size'] for node in tree if node['type'] == 'dir' and node['size'] + unused >= 30000000 ) # search for dirs, that - upon deletion - would free enough memory size for needed, choose minimum
```

# --- Day 8: Treetop Tree House ---

```python
puzzle = '''30373
25512
65332
33549
35390'''
with open('8.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = list(map(list, puzzle))

import numpy as np
puzzle = np.array(puzzle, int)
puzzle

################################################################################
# PART 1

size = puzzle.shape

v = 2 * (sum(size) - 2)
for i in range(1, size[0] - 1):
    for j in range(1, size[1] - 1):
        t = puzzle[i, j]
        v += all(puzzle[:i, j] < t) or all(puzzle[i + 1:, j] < t) or all(puzzle[i, :j] < t) or all(puzzle[i, j + 1:] < t)
v

################################################################################
# PART 2

S = 0
for i in range(1, size[0] - 1): # the scenic score of a tree on sides would be always 0, because of multiplying values
    for j in range(1, size[1] - 1): # enough to select from not-side-trees

        # tree height under consideration
        t = puzzle[i, j]
        # print(i, j, t)

        # logic in each direction:
        # 1. create array of trees going from current to distant (need to reverse in some cases)
        # 2. find the first tree which is as tall as the current
        # 3. calculate distance and multiply
        s = 1

        # up
        u = puzzle[:i, j][::-1] # need to revese, so it goes bottom-up
        m = 1
        if len(u) >= 2:
            while m < len(u) and u[m - 1] < t:
                m += 1
        # print('u', m)
        s *= m

        # down
        d = puzzle[i+1:, j]
        m = 1
        if len(d) >= 2:
            while m < len(d) and d[m - 1] < t:
                m += 1
        # print('d', m)
        s *= m

        # left
        l = puzzle[i, :j][::-1] # need to reverse, so it goes right-to-left
        m = 1
        if len(l) >= 2:
            while m < len(l) and l[m - 1] < t:
                m += 1
        # print('l', m)
        s *= m

        # right
        r = puzzle[i, j+1:]
        m = 1
        if len(r) >= 2:
            while m < len(r) and r[m - 1] < t:
                m += 1
        # print('r', m)
        s *= m

        # maximum search for the highest product
        # print(s)
        if S < s: S = s

S
```

# --- Day 9: Rope Bridge ---

```python
puzzle = '''R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2''' # part 1 = 13, part 2 = 2
puzzle = '''R 5
U 8
L 8
D 3
R 17
D 10
L 25
U 20''' # part 2 = 36
with open('9.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ step.split(' ') for step in puzzle ]
puzzle = [ [ step[0], int(step[1]) ] for step in puzzle  ]
puzzle

################################################################################
# PART 1

H = (0, 0)
T = (0, 0)

visited = [ (0, 0) ]

for step in puzzle:

    dir = None
    if step[0] == 'L': dir = (-1, 0)
    if step[0] == 'R': dir = (1, 0)
    if step[0] == 'U': dir = (0, 1)
    if step[0] == 'D': dir = (0, -1)

    for i in range(step[1]):
        H = (H[0] + dir[0], H[1] + dir[1])
        
        # the tail must follow the head
        # diff shows the H's relative position to T
        diff = (H[0] - T[0], H[1] - T[1])

        if diff == (0, 2):
            T = (T[0], T[1] + 1)
        if diff == (0, -2):
            T = (T[0], T[1] - 1)
        if diff == (2, 0):
            T = (T[0] + 1, T[1])
        if diff == (-2, 0):
            T = (T[0] - 1, T[1])
        if diff == (2, 1) or diff == (1, 2):
            T = (T[0] + 1, T[1] + 1)
        if diff == (-2, 1) or diff == (-1, 2):
            T = (T[0] - 1, T[1] + 1)
        if diff == (-2, -1) or diff == (-1, -2):
            T = (T[0] - 1, T[1] - 1)
        if diff == (2, -1) or diff == (1, -2):
            T = (T[0] + 1, T[1] - 1)

        # print('H:', H, 'T:', T)

        if T not in visited: visited.append(T)

len(visited)

################################################################################
# PART 2

knots = [ (0, 0) for i in range(10) ]
knots

visited = [ (0, 0) ]

for step in puzzle:

    dir = None
    if step[0] == 'L': dir = (-1, 0)
    if step[0] == 'R': dir = (1, 0)
    if step[0] == 'U': dir = (0, 1)
    if step[0] == 'D': dir = (0, -1)

    for i in range(step[1]):
        # print('i:', i)
        H = knots[0]
        H = (H[0] + dir[0], H[1] + dir[1])
        knots[0] = H
        
        # the tail must follow the head
        # diff shows the knot[i - 1]'s relative position to knot[i]
        for j in range(1, 10):
            # print('j:', j)
            H = knots[j - 1]
            T = knots[j]
            diff = (H[0] - T[0], H[1] - T[1])

            if diff == (0, 0):
                break
            if diff == (0, 2):
                T = (T[0], T[1] + 1)
            if diff == (0, -2):
                T = (T[0], T[1] - 1)
            if diff == (2, 0):
                T = (T[0] + 1, T[1])
            if diff == (-2, 0):
                T = (T[0] - 1, T[1])
            # had to add 2-diag-diff to the last 4 cases
            if diff == (2, 1) or diff == (1, 2) or diff == (2, 2):
                T = (T[0] + 1, T[1] + 1)
            if diff == (-2, 1) or diff == (-1, 2) or diff == (-2, 2):
                T = (T[0] - 1, T[1] + 1)
            if diff == (-2, -1) or diff == (-1, -2) or diff == (-2, -2):
                T = (T[0] - 1, T[1] - 1)
            if diff == (2, -1) or diff == (1, -2) or diff == (2, -2):
                T = (T[0] + 1, T[1] - 1)

            knots[j] = T

        # print(knots)

        if knots[-1] not in visited: visited.append(knots[-1])

len(visited)
```

# --- Day 11: Monkey in the Middle ---

```python
puzzle = '''Monkey 0:
  Starting items: 79, 98
  Operation: new = old * 19
  Test: divisible by 23
    If true: throw to monkey 2
    If false: throw to monkey 3

Monkey 1:
  Starting items: 54, 65, 75, 74
  Operation: new = old + 6
  Test: divisible by 19
    If true: throw to monkey 2
    If false: throw to monkey 0

Monkey 2:
  Starting items: 79, 60, 97
  Operation: new = old * old
  Test: divisible by 13
    If true: throw to monkey 1
    If false: throw to monkey 3

Monkey 3:
  Starting items: 74
  Operation: new = old + 3
  Test: divisible by 17
    If true: throw to monkey 0
    If false: throw to monkey 1'''
with open('11.txt') as f: puzzle = f.read()


puzzle = puzzle.split('\n')
puzzle = [ line.strip() for line in puzzle ]
puzzle

# def processing puzzle input, as function, so its reusable, same as day 5
def process_puzzle(puzzle):

    monkeys = []

    i = 0
    while i < len(puzzle):

        new_m = {}

        #
        items = puzzle[i + 1].replace('Starting items: ', '')
        items = items.split(', ')
        items = list(map(int, items))
        new_m['items'] = items

        # monkey inspection operation
        new_m['op'] = puzzle[i + 2].replace('Operation: new = ', '')
        new_m['op'] = new_m['op'].split(' ')[1:]
        if new_m['op'][-1].isnumeric():
            new_m['op'][-1] = int(new_m['op'][-1])


        new_m['mod'] = int(puzzle[i + 3].replace('Test: divisible by ', ''))
        new_m[True] = int(puzzle[i + 4].split(' ')[-1])
        new_m[False] = int(puzzle[i + 5].split(' ')[-1])
            
        # item-inspection-counter = stores puzzle answer
        new_m['IC'] = 0
        monkeys.append(new_m)
        i += 7

    return monkeys

# process
monkeys = process_puzzle(puzzle)
monkeys


################################################################################
# PART 1

for round in range(20):

    # going through each monkey ina round
    for m in monkeys:
        for item in m['items']:

            # monkey inspects item
            m['IC'] += 1
            tmp = item if m['op'][1] == 'old' else m['op'][1]
            if m['op'][0] == '+':
                item += tmp
            else:
                item *= tmp
            item //= 3

            # monkey tests item then throws
            rem = item % m['mod']
            i = m[ rem == 0 ]
            monkeys[i]['items'].append(item)
        
        m['items'] = []

monkeys

# answer
MB = sorted( m['IC'] for m in monkeys )
MB[-2] * MB[-1]

################################################################################
# PART 2

# reset
monkeys = process_puzzle(puzzle)
monkeys

# main modulo
M = 1
for m in monkeys: M *= m['mod']
M


for round in range(10000): # iterating much longer

    # monkey inspects item
    for m in monkeys:
        for item in m['items']:

            # monkey inspects item
            m['IC'] += 1
            tmp = item if m['op'][1] == 'old' else m['op'][1]
            if m['op'][0] == '+':
                item += tmp
            else:
                item *= tmp
            
            # no longer dividing by 3 makes numbers explode
            # however, each monkey has a prime divisor test => we can keep the values uniquely in the product of these prime moduli
            item %= M

            # monkey tests item then throws
            rem = item % m['mod']
            i = m[ rem == 0 ]
            monkeys[i]['items'].append(item)
        
        m['items'] = []

# answer
MB = sorted( m['IC'] for m in monkeys )
MB[-2] * MB[-1]
```

# --- Day 24: Blizzard Basin ---

```python
puzzle = '''#.######
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
######.#'''
with open('24.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = list(map(list, puzzle))
puzzle

R = len(puzzle)
C = len(puzzle[0])

# print('R:', R, 'C:', C)

################################################################################
# PART 1-2

# break up blizzards' movements into states
init_state = {}
for r in range(1, R - 1):
    for c in range(1, C - 1):
        if puzzle[r][c] != '.':

            pos = (r, c)
            dir = puzzle[r][c]            
            if pos not in init_state:
                init_state[pos] = [dir]
            else:
                init_state[pos].append(dir)
init_state

# collect all the states / positions that can occur through blizzards' movements
states = [ init_state ]
while True:
    current = states[-1]
    next = {}

    for pos in current:
        for dir in current[pos]:

            if dir == '>':
                next_pos = ( pos[0], pos[1] + 1 )
            elif dir == '<':
                next_pos = ( pos[0], pos[1] - 1 )
            elif dir == '^':
                next_pos = ( pos[0] - 1, pos[1] )
            else:
                assert dir == 'v'
                next_pos = ( pos[0] + 1, pos[1] )
            
            if next_pos[1] == C - 1:
                next_pos = ( next_pos[0], 1 )
            if next_pos[1] == 0:
                next_pos = ( next_pos[0], C - 2 )
            if next_pos[0] == 0:
                next_pos = ( R - 2, next_pos[1] )
            if next_pos[0] == R - 1:
                next_pos = ( 1, next_pos[1] )

            if next_pos not in next:
                next[ next_pos ] = [dir]
            else:
                next[ next_pos ].append(dir)
    
    if next in states:
        break
    else:
        states.append(next)

S = len(states)
S


# now build a graph from states
graph = {}
for i in range(len(states)):
    s_i = states[i]

    graph[ (0, 1, i) ] = []
    graph[ (R - 1, C - 2, i) ] = []
    for r in range(1, R - 1):
        for c in range(1, C - 1):
            if (r, c) not in s_i:
                graph[ (r, c, i) ] = []

# a node is neighbour to another, if they are one block away, and time increases by 1
for (r, c, i) in graph:
    j = (i + 1) % S
    neighbours = [ (r, c, j), (r, c - 1, j), (r, c + 1, j), (r - 1, c, j), (r + 1, c, j)  ]
    neighbours = [ n for n in neighbours if n in graph ]
    graph[ (r, c, i) ] = neighbours

graph

# apply good ol' Dijkstra
def Dijkstra(start_state, goal_pos):
    global graph

    to_explore = [ start_state ]

    distances = { (r, c, i) : 10 ** 6 for r in range(1, R - 1) for c in range(1, C - 1) for i in range(S) }
    for i in range(S): distances[ (start_state[0], start_state[1], i) ] = 10 ** 6
    for i in range(S): distances[ (goal_pos[0], goal_pos[1], i) ] = 10 ** 6
    distances[ start_state ] = 0

    iter = 0
    while len(to_explore) != 0:
        # print(f'\nIter: {iter}; to explore: {to_explore}')

        current = to_explore.pop(0)
        # print(f'Current: {current}')
        (r, c, i) = current

        if (current[0], current[1]) == goal_pos:
            # print(f'Breaking out after {iter} iterations.')
            break

        neighbours = graph[ current ]
        # print(f'Neighbours: {neighbours}')

        for n in neighbours:
            (x, y, j) = n
            assert (i + 1) % S == j

            d = distances[ current ] + 1
            if d < distances[n]:
                distances[n] = d

                if n not in to_explore:
                    to_explore.append(n)

        iter += 1
        if iter == 10 ** 6:
            break
    
    return distances, current


# PART 1 - one trip to bottom
distances, current_state = Dijkstra( (0, 1, 0), (R - 1, C - 2) )
min( distances[ (R - 1, C - 2, i) ] for i in range(S) )

# PART 2 - to bottom, to back, then again to bottom
total = 0

distances, current_state = Dijkstra( (0, 1, 0), (R - 1, C - 2) )
total += min( distances[ (R - 1, C - 2, i) ] for i in range(S) )
# print(f'To goal: {total}')

distances, current_state = Dijkstra( current_state, (0, 1) )
total += min( distances[ (0, 1, i) ] for i in range(S) )
# print(f'To goal: {total}')

distances, current_state = Dijkstra( current_state, (R - 1, C - 2) )
total += min( distances[ (R - 1, C - 2, i) ] for i in range(S) )
total
```
