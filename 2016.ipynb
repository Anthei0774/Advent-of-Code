# --- Day 2: Bathroom Security ---

```python
puzzle = '''ULL
RRDDD
LURDL
UUUUD'''
with open('2.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle

################################################################################
# PART 1

numpad = [['1', '2', '3'],
          ['4', '5', '6'],
          ['7', '8', '9']]
r = 1
c = 1

code = ''
for instruction in puzzle:
    for char in instruction:

        if char == 'L':
            c = 0 if c in [0, 1] else 1
        elif char == 'R':
            c = 2 if c in [1, 2] else 1
        elif char == 'U':
            r = 0 if r in [0, 1] else 1
        else:
            assert char == 'D'
            r = 2 if r in [1, 2] else 1

    code += numpad[r][c]

code

################################################################################
# PART 2

numpad = [['-', '-', '1', '-', '-'], #           0,2
          ['-', '2', '3', '4', '-'], #      1,1  1,2  1,3
          ['5', '6', '7', '8', '9'], # 2,0  2,1  2,2  2,3  2,4
          ['-', 'A', 'B', 'C', '-'], #      3,1  3,2  3,3
          ['-', '-', 'D', '-', '-']] #           4,2
r = 2
c = 0

code = ''
for instruction in puzzle:
    for char in instruction:

        if char == 'L':
            if (r == 2 and c in [1, 2, 3, 4]) or (r in [1, 3] and c in [2, 3]):
                c -= 1
        elif char == 'R':
            if (r == 2 and c in [0, 1, 2, 3]) or (r in [1, 3] and c in [1, 2]):
                c += 1
        elif char == 'U':
            if (c == 2 and r in [1, 2, 3, 4]) or (c in [1, 3] and r in [2, 3]):
                r -= 1
        else:
            assert char == 'D'
            if (c == 2 and r in [0, 1, 2, 3]) or (c in [1, 3] and r in [1, 2]):
                r += 1

    code += numpad[r][c]

code
```

# --- Day 4: Security Through Obscurity ---

```python
puzzle = '''aaaaa-bbb-z-y-x-123[abxyz]
a-b-c-d-e-f-g-h-987[abcde]
not-a-real-room-404[oarel]
totally-real-room-200[decoy]'''
puzzle = 'qzmt-zixmtkozy-ivhz-343[zimth]'
with open('4.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, line in enumerate(puzzle):
    line = line.split('-')

    room = {}
    room['name'] = '-'.join(line[:-1])
    room['sectorid'] = int(line[-1].split('[')[0])
    room['checksum'] = line[-1].split('[')[-1][:-1]
    puzzle[i] = room

puzzle

################################################################################
# PART 1

roomLetters = []
for room in puzzle:
    letterCounts = []

    # count letters in room name
    for char in set(room['name']):
        if char == '-': continue
        letterCounts.append([room['name'].count(char), char])

    letterCounts = sorted(letterCounts, key = lambda x: (-x[0], x[1]) ) # sort by numbers descending, then letters ascending
    letterCounts = letterCounts[:5] # take top 5 most occuring
    roomLetters.append(letterCounts)

# concatenate the 5 letters into a single word
roomLetters = [ ''.join([ letterCountPair[1] for letterCountPair in room ]) for room in roomLetters ]
roomLetters


s = 0
for i, rL in enumerate(roomLetters):
    room = puzzle[i]
    room['decoy'] = True

    # room is not a decoy
    if rL == room['checksum']:
        room['decoy'] = False
        s += room['sectorid']
s

################################################################################
# PART 2

abc = 'abcdefghijklmnopqrstuvwxyz'

for room in puzzle:
    result = ''

    if not room['decoy']:
        for char in room['name']:

            if char == '-':
                result += '-'
                continue 
            i = abc.index(char)
            i = (i + room['sectorid']) % 26
            result += abc[i]

        if 'northpole' in result:
            break

room['sectorid']
```

# --- Day 5: How About a Nice Game of Chess? ---

```python
puzzle = 'abc'
puzzle = 'abbhdwsy' # puzzle input

################################################################################
# PART 1

import hashlib
password = ''

i = 0
for iter in range(8):

    while True:
        i += 1

        string = puzzle + str(i)
        hash = hashlib.md5(string.encode())
        hex = hash.hexdigest()
        if hex[:5] == '00000': break

    password += hex[5]

password

################################################################################
# PART 2

password = ['', '', '', '', '', '', '', '']

i = 0
iter = 0
while iter < 8:

    while True:
        i += 1
        string = puzzle + str(i)
        hash = hashlib.md5(string.encode())
        hex = hash.hexdigest()
        if hex[:5] == '00000' and hex[5].isnumeric():
            idx = int(hex[5])
            if idx in range(8):    
                break

    if password[idx] == '':
        password[idx] = hex[6]    
        iter += 1

password = ''.join(password)
password
```

# --- Day 7: Internet Protocol Version 7 ---

```python
# part 1 example
puzzle = '''abba[mnop]qrst
abcd[bddb]xyyx
aaaa[qwer]tyui
ioxxoj[asdfgh]zxcvbn'''

# part 2 example
puzzle = '''aba[bab]xyz
xyx[xyx]xyx
aaa[kek]eke
zazbz[bzb]cdb'''

with open('7.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle

################################################################################
# PART 1

supports_tls = 0

for ip in puzzle:
    # print(ip)

    start = ip.find('[')
    end = ip.find(']')

    # check interior, if has abba, then skip
    has_abba = False
    for i in range(start + 1, end - 3):
        seq = ip[i : i + 4]
        if len(set(seq)) == 2:
            if seq[0] == seq[3]:
                # print('Invalid interior')
                has_abba = True
                break

    if has_abba: continue


    # check prefix, if has abba, increment tls counter, then skip
    for i in range(start - 3):
        seq = ip[i : i + 4]
        if len(set(seq)) == 2:
            if seq[0] == seq[3]:
                # print('Prefix has abba')
                has_abba = True
                supports_tls += 1
                break
                
    if has_abba: continue
    

    # check suffix
    for i in range(end + 1, len(ip) - 3):
        seq = ip[i : i + 4]
        if len(seq) < 4: break
        if len(set(seq)) == 2:
            if seq[0] == seq[3]:
                # print('Suffix has abba')
                has_abba = True
                supports_tls += 1
                continue

supports_tls

################################################################################
# PART 2

def extract_3char(string):
    trios = []
    for i in range(len(string) - 2):
        sub = string[i : i + 3]
        if sub[0] == sub[2] and sub[0] != sub[1] and sub[1] not in ['[', ']']:
            if sub not in trios:
                trios.append(sub)
    return trios


# find all trios inside brackets
inside_brackets = []

for ip in puzzle:
    # print(ip)

    all_trios = []
    start = -1
    while True:

        start = ip.find('[', start + 1)
        if start == -1: break
        end = ip.find(']', start + 1)
        trios = extract_3char( ip[start + 1 : end] )
        all_trios += trios

    inside_brackets.append(all_trios)


# find all trios outside brackets
outside_brackets = []
for ip in puzzle:
    # print(ip)

    all_trios = []
    start = -1
    end = 0
    while True:

        start = ip.find('[', end + 1)

        if start != -1:
            trios = extract_3char( ip[end : start] )
            all_trios += trios
            end = ip.find(']', start + 1)
        else:
            trios = extract_3char( ip[end : len(ip)] )
            all_trios += trios
            break

    all_trios = list(set(all_trios))
    outside_brackets.append(all_trios)


# match inside-outside
supports_ssl = 0
for i in range(len(puzzle)):

    
    for trio in inside_brackets[i]:
        if trio[1] + trio[0] + trio[1] in outside_brackets[i]:
            # print(puzzle[i], trio)
            supports_ssl += 1

            # break out to avoid counting multiple matching trios and thus giving more points for an IP
            break 

supports_ssl
```

```python
puzzle = '''rect 3x2
rotate column x=1 by 1
rotate row y=0 by 4
rotate column x=1 by 1'''
with open('8.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
for i, line in enumerate(puzzle):
    # print(line)

    new_line = {}
    new_line['command'] = line.split(' ')[0]

    if new_line['command'] == 'rect':
        sides = line[5:].split('x')
        [col, row] = list(map(int, sides))
        new_line['col'] = col
        new_line['row'] = row
    
    else:
        new_line['axis'] = ( line.split(' ')[1], int(line.split(' ')[2].split('=')[-1]) )
        new_line['shift'] = int(line.split(' ')[-1])
        
    # print(new_line)
    puzzle[i] = new_line

################################################################################
# PART 1

import numpy as np

# example
# screen = [ [ 0 for col in range(7) ] for row in range(3) ]

# 50 * 6 screen
screen = [ [ 0 for col in range(50) ] for row in range(6) ]

screen = np.array(screen)

for line in puzzle:
    # print('\n', line)

    if line['command'] == 'rect':
        row = line['row']
        col = line['col']
        screen[ : row, : col] = 1
    
    # rotate
    # useful stuff: https://numpy.org/doc/stable/reference/generated/numpy.roll.html
    else:

        if line['axis'][0] == 'row':
            r = line['axis'][1]
            s = line['shift']

            screen[r] = np.roll(screen[r], s)

        else:
            c = line['axis'][1]
            s = line['shift']

            screen[:, c] = np.roll(screen[:, c], s)

    # print
    # tmp = screen.tolist()
    # tmp = [ list(map(str, line)) for line in tmp]
    # for line in tmp: print(''.join(line))

np.sum(screen)

################################################################################
# PART 2

for col in range(0, 50, 5):
    tmp = screen[:, col : col + 5].tolist()
    tmp = [ list(map(str, line)) for line in tmp ]
    tmp = [ ''.join(line) for line in tmp ]
    tmp = '\n'.join(tmp)
    tmp = tmp.replace('0', '.').replace('1', '#')
    print(tmp, '\n')
```

# --- Day 9: Explosives in Cyberspace ---

```python
# part 1 example
puzzle = '''ADVENT
A(1x5)BC
(3x3)XYZ
A(2x2)BCD(2x2)EFG
(6x1)(1x3)A
X(8x2)(3x3)ABCY'''

# part 2 example
puzzle = '''(3x3)XYZ
X(8x2)(3x3)ABCY
(27x12)(20x12)(13x14)(7x10)(1x12)A
(25x3)(3x3)ABC(2x3)XY(5x2)PQRSTX(18x9)(3x2)TWO(5x7)SEVEN'''

with open('9.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle

################################################################################
# PART 1

def decompress_line_part1(line):
    # print('Line:', line)

    # find first parentheses
    start = line.find('(')
    if start != -1:
        end = line.find(')', start + 1)

        prefix = line[: start]
        # print('Prefix:', prefix)

        # extract sequence length and repetition times from the parenthesis
        [length, repeat] = list(map(int, line[start + 1 : end].split('x')))


        suffix = line[end + 1 + length :]
        suffix = decompress_line_part1(suffix)
        # print('Suffix:', suffix)

        repeated = line[end + 1 : end + 1 + length] * repeat
        # print('Repeated:', repeated)

        new_line = prefix + repeated + suffix

        return new_line

    # no parentheses = nothing to decompress
    else:
        # print('No decompression')
        return line


decompressed = []

for line in puzzle:
    new_line = decompress_line_part1(line)
    # print('Decompressed line:', new_line)
    # print(len(new_line))
    # print('\n---\n')


################################################################################
# PART 2

def decompress_line_part2(line):
    # print('Line:', line)

    # find first parentheses
    start = line.find('(')
    if start != -1:
        end = line.find(')', start)

        prefix = line[: start]
        l_prefix = len(prefix)
        # print('Prefix:', prefix, '; len:', l_prefix)
        

        # extract sequence length and repetition times from the parenthesis
        [length, repeat] = list(map(int, line[start + 1 : end].split('x')))


        # need to decompress both suffix and region after parenthesis
        # return only the length of the decompressed line

        # 1st suffix
        suffix = line[end + 1 + length :]
        l_suffix = decompress_line_part2(suffix)
        # print('Suffix:', suffix, '; len:', l_suffix)

        # 2nd inside
        inside = line[end + 1 : end + 1 + length]
        l_inside = decompress_line_part2(inside)
        l_inside_repeated = l_inside * repeat
        # print('Inside:', inside, '; len:', l_inside, '; repeated len:', l_inside_repeated)

        # sum up
        l_total = l_prefix + l_inside_repeated + l_suffix
        return l_total

    # no parentheses = nothing to decompress
    else:
        # print('No decompression')
        return len(line)


for line in puzzle:
    l_total = decompress_line_part2(line)
    print('Total decompression len:', l_total)
    # print('\n---\n')
```

# --- Day 10: Balance Bots ---

```python
puzzle = '''value 5 goes to bot 2
bot 2 gives low to bot 1 and high to bot 0
value 3 goes to bot 1
bot 1 gives low to output 1 and high to bot 0
bot 0 gives low to output 2 and high to output 0
value 2 goes to bot 2'''

with open('10.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')

# sort first by bot / value, then by index
puzzle = sorted(puzzle, key = lambda x: (x.split(' ')[0], int(x.split(' ')[1])) )
puzzle

bots = []

for line in puzzle:
    # print(line)

    line_type = line.split(' ')[0]
    if line_type == 'bot':
        bot = {
            'low': int(line.split(' ')[6])
            , 'low_bot': line.split(' ')[5] == 'bot'
            , 'high': int(line.split(' ')[-1])
            , 'high_bot': line.split(' ')[-2] == 'bot'
            , 'values': []
            }

        # print(bot)
        bots.append(bot)

    else:
        value = int(line.split(' ')[1])
        index = int(line.split(' ')[-1])

        bots[index]['values'].append(value)
        bots[index]['values'].sort()

bots

################################################################################
# PART 1

outputs = {}

found_bot = False

# iterate until the bot found with [17, 61] ready to be given to other bots / outputs
iter = 0
while not found_bot:
    # print('Iteration:', iter)
    iter += 1

    givers = [ (i, b) for i, b in enumerate(bots) if len(b['values']) == 2 ]

    for (i, b) in givers:
        # print(b)

        if b['values'] == [17, 61]:
            found_bot = True
            break

        # if low target is bot => pass values, then sort
        if b['low_bot']:
            bots[ b['low'] ]['values'].append(b['values'][0])
            bots[ b['low'] ]['values'].sort()
        # low target is output => initiate new output channel or append to the end
        else:
            if b['low'] not in outputs:
                outputs[b['low']] = [b['values'][0]]
            else:
                outputs[b['low']].append(b['values'][0])

        # same logic on high end
        if b['high_bot']:
            bots[ b['high'] ]['values'].append(b['values'][1])
            bots[ b['high'] ]['values'].sort()
        else:
            if b['high'] not in outputs:
                outputs[b['high']] = [b['values'][1]]
            else:
                outputs[b['high']].append(b['values'][1])

        # empty bot values
        b['values'] = []

i

################################################################################
# PART 2

# continue iterating but with a different stopping critera (= no more moves)
iter = 0
while True:
    # print('Iteration:', iter)
    iter += 1

    givers = [(i, b) for i, b in enumerate(bots) if len(b['values']) == 2]
    if givers == []: break # stable state of bots, no values passed

    for (i, b) in givers:
        # print(b)

        # if low target is bot => pass values, then sort
        if b['low_bot']:
            bots[ b['low'] ]['values'].append(b['values'][0])
            bots[ b['low'] ]['values'].sort()
        # low target is output => initiate new output channel or append to the end
        else:
            if b['low'] not in outputs:
                outputs[b['low']] = [b['values'][0]]
            else:
                outputs[b['low']].append(b['values'][0])

        # same logic on high end
        if b['high_bot']:
            bots[ b['high'] ]['values'].append(b['values'][1])
            bots[ b['high'] ]['values'].sort()
        else:
            if b['high'] not in outputs:
                outputs[b['high']] = [b['values'][1]]
            else:
                outputs[b['high']].append(b['values'][1])

        # empty bot values
        b['values'] = []


outputs[0][0] * outputs[1][0] * outputs[2][0]
```

# --- Day 11: Radioisotope Thermoelectric Generators ---

```python
# example input
'''The first floor contains a hydrogen-compatible microchip and a lithium-compatible microchip.
The second floor contains a hydrogen generator.
The third floor contains a lithium generator.
The fourth floor contains nothing relevant.'''

# puzzle input 
'''The first floor contains a thulium generator, a thulium-compatible microchip, a plutonium generator, and a strontium generator.
The second floor contains a plutonium-compatible microchip and a strontium-compatible microchip.
The third floor contains a promethium generator, a promethium-compatible microchip, a ruthenium generator, and a ruthenium-compatible microchip.
The fourth floor contains nothing relevant.'''

# adding these to the first floor as part 2 puzzle input
'''An elerium generator.
An elerium-compatible microchip.
A dilithium generator.
A dilithium-compatible microchip.'''




# encoding the inputs by hand, start and end states
s = '0,HydrM-LithM,HydrG,LithG,'
e = '3,,,,HydrG-HydrM-LithG-LithM'

s = '0,PlutG-StroG-ThulG-ThulM,PlutM-StroM,PromG-PromM-RuthG-RuthM,'
e = '3,,,,PlutG-PlutM-PromG-PromM-RuthG-RuthM-StroG-StroM-ThulG-ThulM'

s = '0,DiliG-DiliM-ElerG-ElerM-ThulG-ThulM-PlutG-StroG,PlutM-StroM,PromG-PromM-RuthG-RuthM,'
e = '3,,,,DiliG-DiliM-ElerG-ElerM-PlutG-PlutM-PromG-PromM-RuthG-RuthM-StroG-StroM-ThulG-ThulM'

states = {}
states[s] = False


################################################################################
# PART 1-2


def validate_state(state):

    # transform string representation of state in list of lists
    proc_state = state.split(',')
    assert len(proc_state) == 5

    proc_state[0] = int(proc_state[0])
    proc_state[1 :] = [ l.split('-') for l in proc_state[1 :] ]

    # iterate over levels of the state
    for level in proc_state[1 :]:
        # and over items on levels
        for item in level:
            
            # for every Microchip, two cases are valid
            # 1. no Generators on its level
            # 2. if there are Generators, then the level must contain the macthing Generator
            if 'M' in item:
                Gs_on_level = [ item for item in level if item[-1] == 'G' ]

                if len(Gs_on_level) != 0:
                    G_matching_to_M = item[: -1] + 'G'
                    if G_matching_to_M not in Gs_on_level:
                        return False
    
    return True




# implementing a Breadth First Search algorithm
iter = 0
while True:
    print('Iteration:', iter)

    # gather those states which havent been checked so far
    not_processed_states = [ s for s in states if not states[s] ]
    if not_processed_states == []: break

    # go through those unchecked states
    for nps in not_processed_states:

        # processing state into list format
        states[nps] = True
        # print('A non-processed state:', nps)
        level_items = nps.split(',')[1 :]
        level_items = [ level.split('-') if level != '' else [] for level in level_items ]
        # print('Levels with items:', level_items)
        elevator_level = int(nps.split(',')[0])
        # print('Elevator level:', elevator_level)
        elevator_level_items = level_items[elevator_level]
        # print('Elevator level items:', elevator_level_items)

        # the elevator will bring up 1 or 2 items
        item_combinations = []
        for i in range(len(elevator_level_items)):
            item_combinations.append([elevator_level_items[i]])
            for j in range(i + 1, len(elevator_level_items)):
                item_combinations.append([elevator_level_items[i], elevator_level_items[j]])
        # print('Item combinations taken up / down:', item_combinations)


        # decide directions: elevators goes up if not on F4 and down if not on F1
        directions = []
        if elevator_level != 3: directions.append(1)
        if elevator_level != 0: directions.append(-1)


        # construct new states from directions and item combinations
        new_states = []
        for d in directions:
            for ic in item_combinations:
                # print('Combination direction + items:', 'up' if d == 1 else 'down', ',', ic)

                # start constructing new state by adding elevator level
                elevator_target_level = elevator_level + d
                ns = str(elevator_target_level)

                # iterate through level
                for i in range(4):
                    # remove items from elevator level
                    if i == elevator_level:
                        remaining_items_on_level = [ item for item in level_items[i] if item not in ic ]
                        # print('Remaining items on level:', remaining_items_on_level)
                        
                        ns += (',' + '-'.join(remaining_items_on_level))
                        continue

                    # add items to elevator target level
                    if i == elevator_target_level:
                        new_items_on_level = level_items[i] + ic
                        new_items_on_level = sorted(new_items_on_level)
                        # print('New items on level:', new_items_on_level)

                        ns += (',' + '-'.join(new_items_on_level))
                        continue

                    # copy the level otherwise
                    ns += (',' + '-'.join(level_items[i]))
                
                # state constructed, append to new if not a previously generated state (= not allowing a backstep)
                if ns not in states:
                    new_states.append(ns)


        # validate & insert new states
        for ns in new_states:
            if validate_state(ns):
                states[ns] = False

    
    # break out if target state reached
    # iteration number + 1 is the steps needed
    print('Total number of states:', len(states))
    if e in states:
        break
    
    # failsafe break to not loop endlessly
    iter += 1
    if iter == 1000:
        break

# final answer
iter + 1
```

```python
def poly(x, y): return x*x + 3*x + 2*x*y + y + y*y

puzzle = 10
size = (7, 10)

puzzle = 1362 # puzzle input
size = (50, 50) # setting it big enough for part 1
# size = (100, 100) # setting it bigger for part 2

grid = [ [ poly(x, y) + puzzle for x in range(size[1]) ] for y in range(size[0]) ] # generate grid with poly and office designer's favourite number
grid = [ [ format(num, 'b').count('1') for num in line ] for line in grid ] # into binary, then count ones
grid = [ [ int(num % 2 == 1) for num in line ] for line in grid ]
grid


################################################################################
# PART 1-2

reached_points = {
    (1, 1) : {'checked': False, 'cost' : 0}
}

# main idea to implement a breadth first search - finding shortest path to any target

# fail switch if while-loop goes out of hand
iter = 0
while True:
    
    all_new_points = []

    # iterate through reached points, which are not checked for neighbours
    for coords in reached_points:
        rp = reached_points[coords]
        if not rp['checked']:
            rp['checked'] = True

            # find valid neighbours
            new_points = [(coords[0] + 1, coords[1]), (coords[0] - 1, coords[1]), (coords[0], coords[1] + 1), (coords[0], coords[1] - 1)] # step left-right-up-down
            new_points = [ np for np in new_points if np[0] >= 0 and np[0] < size[1] and np[1] >= 0 and np[1] < size[0] and grid[np[1]][np[0]] == 0 ] # staying on the grid and not wall
            # print(new_points)

            # save parent and list of points reached from there
            all_new_points.append({ 'parent': rp, 'new_points': new_points })

    
    # no new points = maze explored
    if all_new_points == []:
        break
    # print(all_new_points)


    # for each new point, append to reached (if not already in) with cost increased
    for a_np in all_new_points:
        for np in a_np['new_points']:
            if np not in reached_points:
                reached_points[np] = {'checked': False, 'cost' : rp['cost'] + 1}


    # failsafe break
    if iter == 1000:
        break
    iter += 1


reached_points[ (7, 4) ] # example answer
reached_points[ (31, 39) ] # part 1 answer
len({ coords : reached_points[coords] for coords in reached_points if reached_points[coords]['cost'] <= 50 }) # part 2 answer
```

# --- Day 15: Timing is Everything ---

```python
# The example problem can be formulated as follows:
# t + 1 + 4 = 0 mod(5)   <===>   t == 0 mod(5)
# t + 2 + 1 = 0 mod(2)   <===>   t == 1 mod(2)

# example input
puzzle = '''Disc #1 has 5 positions; at time=0, it is at position 4.
Disc #2 has 2 positions; at time=0, it is at position 1.'''

discs = [
    {'num_pos' : 5, 'start' : 4}
    , {'num_pos' : 2, 'start' : 1}
]

remainders = [0, 1]


# For the puzzle input:
# t + 1 +  2 = 0 mod( 5)   <===>   t = 2 mod( 5)
# t + 2 +  7 = 0 mod(13)   <===>   t = 4 mod(13)
# t + 3 + 10 = 0 mod(17)   <===>   t = 4 mod(17)
# t + 4 +  2 = 0 mod( 3)   <===>   t = 0 mod( 3)
# t + 5 +  9 = 0 mod(19)   <===>   t = 5 mod(19)
# t + 6 +  0 = 0 mod( 7)   <===>   t = 1 mod( 7)
# t + 7 +  0 = 0 mod(11)   <===>   t = 4 mod(11)

puzzle = '''Disc #1 has 5 positions; at time=0, it is at position 2.
Disc #2 has 13 positions; at time=0, it is at position 7.
Disc #3 has 17 positions; at time=0, it is at position 10.
Disc #4 has 3 positions; at time=0, it is at position 2.
Disc #5 has 19 positions; at time=0, it is at position 9.
Disc #6 has 7 positions; at time=0, it is at position 0.'''

discs = [
    {'num_pos' : 5, 'start' : 2}
    , {'num_pos' : 13, 'start' : 7}
    , {'num_pos' : 17, 'start' : 10}
    , {'num_pos' : 3, 'start' : 2}
    , {'num_pos' : 19, 'start' : 9}
    , {'num_pos' : 7, 'start' : 0}
    , {'num_pos' : 11, 'start' : 0} # part 2 tweak: adding a new disc with start pos 0 and 11 pos in total
]

remainders = [
    2
    , 4
    , 4
    , 0
    , 5
    , 1
    , 4 # part 2 tweak reduces to 4 remainder
]

################################################################################
# PART 1-2

# this latter equation system can be solved with chinese remainder theorem - glad I studied math for bachelors
# https://en.wikipedia.org/wiki/Chinese_remainder_theorem#Existence_(direct_construction)
# https://en.wikipedia.org/wiki/Extended_Euclidean_algorithm#Pseudocode


N = 1
for d in discs: N *= d['num_pos']
# print('Product of moduli = N:', N)


def extended_gcd(a, b):
    (old_r, r) = (a, b)
    (old_s, s) = (1, 0)
    (old_t, t) = (0, 1)    
    
    while r != 0:
        q = old_r // r
        (old_r, r) = (r, old_r - q * r)
        (old_s, s) = (s, old_s - q * s)
        (old_t, t) = (t, old_t - q * t)
    
    # print('Bézout coefficients:', (old_s, old_t))
    # print('greatest common divisor:', old_r)
    return (old_s, old_t)
    

calculations = []

for i, d in enumerate(discs):
    ni = d['num_pos']
    Ni = N // ni
    # print('Coprimes:', Ni, ni)

    new_calc = {}
    new_calc['Ni'], new_calc['ni'] = Ni, ni
    new_calc['Mi'], new_calc['mi'] = extended_gcd(Ni, ni)
    new_calc['ai'] = remainders[i]
    # print('Coefficients:', new_calc)
    calculations.append(new_calc)

    # if i == 3: break

calculations

x = 0
for calc in calculations: x += calc['Ni'] * calc['Mi'] * calc['ai']
while x < 0: x += N
x
```

```python
puzzle_input = 'abcde'
puzzle_commands = '''swap position 4 with position 0
swap letter d with letter b
reverse positions 0 through 4
rotate left 1
move position 1 to position 4
move position 3 to position 0
rotate based on position of letter b
rotate based on position of letter d'''

# part 1 puzzle input and commands
puzzle_input = 'abcdefgh'
with open('21.txt') as f:
    puzzle_commands = f.read()

puzzle_input = 'abcdefgh'

puzzle_commands = puzzle_commands.split('\n')

################################################################################
# PART 1

def exec_command(output, c):
    c = c.split(' ')

    if ' '.join(c[: 2]) == 'swap position':
        i = int(c[2])
        j = int(c[-1])

        output = list(output)
        tmp = output[i]
        output[i] = output[j]
        output[j] = tmp
        output = ''.join(output)

    if ' '.join(c[: 2]) == 'swap letter':
        a = c[2]
        b = c[-1]

        output = output.replace(a, '*')
        output = output.replace(b, a)
        output = output.replace('*', b)
    
    if ' '.join(c[: 2]) == 'reverse positions':
        i = int(c[2])
        j = int(c[-1])
        assert i <= j

        output = list(output)
        
        prefix = output[: i]
        mid = output[i : j + 1]
        mid.reverse()
        suffix = output[j + 1 :]
        
        output = prefix + mid + suffix
        output = ''.join(output)

    if c[0] == 'rotate':

        if c[1] == 'based':
            l = c[-1]
            i = output.index(l)
            # print('Index (incremented only once):', i)
            if i >= 4: i += 1
            i += 1
            while i >= len(output): i -= len(output)
            output = output[-i :] + output[: -i]

        else:
            i = int(c[2])
            output = output[i :] + output[: i] if c[1] == 'left' else output[-i :] + output[: -i]

    if c[0] == 'move':
        i = int(c[2])
        j = int(c[-1])
        
        output = list(output)
        tmp = output[i]
        del output[i]

        if j == len(output):
            output.append(tmp)
        else:
            output = output[: j] + [tmp] + output[j :]

        output = ''.join(output)

    return output


output = puzzle_input
# print('Init:', output, '\n')
for c in puzzle_commands:
    # print('Command:', c)
    output = exec_command(output, c)
    # print('After:', output, '\n')
output

################################################################################
# PART 2

def perm(text):
    assert len(text) >= 2

    if len(text) == 2:
        swap = text[1] + text[0]
        return [ text, swap ]

    else:
        final = []
        for i in range(len(text)):
            sub = text[: i] + text[i + 1 :]
            sub = perm(sub)
            sub = [ text[i] + p for p in sub ]
            final += sub
        return final

all_inputs = perm(puzzle_input)


for inp in all_inputs:
    output = inp
    for c in puzzle_commands:
        output = exec_command(output, c)
    if output == 'fbgdceah':
        break

inp
```

# --- Day 22: Grid Computing ---

```python
puzzle = '''root@ebhq-gridcenter# df -h
Filesystem            Size  Used  Avail  Use%
/dev/grid/node-x0-y0   10T    8T     2T   80%
/dev/grid/node-x0-y1   11T    6T     5T   54%
/dev/grid/node-x0-y2   32T   28T     4T   87%
/dev/grid/node-x1-y0    9T    7T     2T   77%
/dev/grid/node-x1-y1    8T    0T     8T    0%
/dev/grid/node-x1-y2   11T    7T     4T   63%
/dev/grid/node-x2-y0   10T    6T     4T   60%
/dev/grid/node-x2-y1    9T    8T     1T   88%
/dev/grid/node-x2-y2    9T    6T     3T   66%'''
th = 30

with open('22.txt') as f:
    puzzle = f.read()
    th = 500

puzzle = puzzle.split('\n')
puzzle = puzzle[2 :]
puzzle = [ line.split(' ') for line in puzzle ]
puzzle = [ [ item for item in line if item != '' ] for line in puzzle ]
for i, line in enumerate(puzzle):
    
    line[0] = line[0].split('-')
    line[0] = line[0][1:]
    line[0] = tuple([ int(item[1:]) for item in line[0] ])
    line[0] = line[0][::-1]

    for k in [1, 2, 3, 4]: line[k] = int(line[k][:-1])
    puzzle[i] = line[:-1]

puzzle = { line[0] : { 'size' : line[1], 'used' : line[2], 'avail' : line[3] } for line in puzzle }
puzzle

N = len(puzzle)

################################################################################
# PART 1

c = 0
for coords_i in puzzle:
    for coords_j in puzzle:
        node_i = puzzle[coords_i]
        node_j = puzzle[coords_j]
        if node_i != node_j:
            c += (node_i['used'] != 0 and node_i['used'] <= node_j['avail'])
c

################################################################################
# PART 2

R = max([ coords[0] for coords in puzzle ])
C = max([ coords[1] for coords in puzzle ])

for coords in puzzle:
    (r, c) = coords
    neighbours = [ (r, c - 1), (r, c + 1), (r - 1, c), (r + 1, c)  ]
    neighbours = [ n for n in neighbours if 0 <= n[0] and n[0] <= R and 0 <= n[1] and n[1] <= C ]
    neighbours = [ n for n in neighbours if puzzle[n]['used'] <= puzzle[coords]['size']  ]
    neighbours = [ n for n in neighbours if puzzle[n]['size'] < th  ]
    puzzle[ coords ]['conn'] = neighbours
puzzle


#
def Dijkstra(source, target, skip = None):
    global puzzle, N

    tmp = { coords : {'dist' : N ** 4, 'prev' : None } for coords in puzzle }
    tmp[source]['dist'] = 0

    to_process = [ coords for coords in tmp if coords != skip ]
    while to_process != []:

        # choose a point with minimal distance from the source
        m = min([ tmp[coords]['dist'] for coords in to_process ])
        next = [ coords for coords in to_process if tmp[coords]['dist'] == m ][0]

        if next == target:
            break

        # process it
        to_process.remove(next)
        for nextnext in puzzle[next]['conn']:
            d = m + 1
            if d < tmp[nextnext]['dist']:
                tmp[nextnext]['dist'] = d
                tmp[nextnext]['prev'] = next

    S = []
    prev = target
    if tmp[prev]['prev'] != None or prev == source:
        while tmp[prev]['prev'] != None:
            S.append(prev)
            prev = tmp[prev]['prev']
    S = S[::-1]
    return S


#
data_path = Dijkstra( source = (0, C), target = (0, 0) )
L = len(data_path)
data_path
print('Data path:', data_path)

#
current = [ coords for coords in puzzle if puzzle[coords]['used'] == 0 ]
assert len(current) == 1
current = current[0]

s = 0
data_pos = (0, C)
for i in range(L):
    data_next = data_path[i]
    
    steps = Dijkstra(source = current, target = data_next, skip = data_pos)
    print(i, steps)
    s += len(steps)

    current = data_pos
    data_pos = data_next
    s += 1

s
```

```python
# example input
puzzle = '''cpy 2 a
tgl a
tgl a
tgl a
cpy 1 a
dec a
dec a'''

# puzzle input
puzzle = '''cpy a b
dec b
cpy a d
cpy 0 a
cpy b c
inc a
dec c
jnz c -2
dec d
jnz d -5
dec b
cpy b c
cpy c d
dec d
inc c
jnz d -2
tgl c
cpy -16 c
jnz 1 c
cpy 81 c
jnz 94 d
inc a
inc d
jnz d -2
inc c
jnz c -5'''

# modified puzzle input
puzzle = '''cpy a b
add b -1
cpy a d
mul d b
cpy d a
cpy c 0
cpy d 0
jnz 0 0
jnz 0 0
jnz 0 0
add b -1
cpy b c
mul c 2
cpy 0 d
jnz 0 0
jnz 0 0
tgl c
cpy -16 c
jnz 1 c
cpy 81 c
jnz 94 d
inc a
inc d
jnz d -2
inc c
jnz c -5'''

puzzle = puzzle.split('\n')
puzzle = [ line.split(' ') for line in puzzle ]
puzzle

################################################################################
# PART 1

registers = {'a' : 0, 'b' : 0, 'c' : 0, 'd' : 0}
registers['a'] = 7 # part 1
registers['a'] = 12 # part 2

from time import sleep

i = 0
iter = 0
while i < len(puzzle):

    # read command and behave appropriately after that, cpy-inc-dec-jnz
    print(i, puzzle[i])
    command = puzzle[i][0]

    if command == 'inc':
        target = puzzle[i][1]
        registers[target] += 1
        i += 1
    
    if command == 'dec':
        target = puzzle[i][1]
        registers[target] -= 1
        i += 1

    if command == 'add':
        target = puzzle[i][1]
        value = puzzle[i][2]
        value = registers[value] if value in registers else int(value)
        registers[target] += value
        i += 1

    if command == 'mul':
        target = puzzle[i][1]
        value = puzzle[i][2]
        value = registers[value] if value in registers else int(value)
        registers[target] *= value
        i += 1

    if command == 'cpy':
        value = puzzle[i][1]
        value = registers[value] if value in registers else int(value)
        target = puzzle[i][2]
        if target in registers:
            registers[target] = value
        i += 1
        
    if command == 'jnz':
        value = puzzle[i][1]
        value = registers[value] if value in registers else int(value)

        if value != 0:
            value = puzzle[i][2]
            value = registers[value] if value in registers else int(value)
            i += value
        else:
            print('Not jumping.')
            i += 1

    if command == 'tgl':
        value = puzzle[i][1]
        value = registers[value] if value in registers else int(value)
        
        target_idx = i + value
        if target_idx < 0 or target_idx >= len(puzzle):
            print('Toggling outside the program.')
            i += 1
            continue
        
        else:
            target = puzzle[target_idx]
            print('Old command:', target)

            if target[0] == 'inc':
                target[0] = 'dec'
                puzzle[target_idx] = target
                i += 1
                continue
            if target[0] in ['dec', 'tgl']:
                target[0] = 'inc'
                puzzle[target_idx] = target
                i += 1
                continue
            if target[0] == 'jnz':
                target[0] = 'cpy'
                puzzle[target_idx] = target
                i += 1
                continue
            if target[0] == 'cpy':
                target[0] = 'jnz'
                puzzle[target_idx] = target
                i += 1
                continue
    
    print('Iter:', iter, '; registers:', registers, '\n')

    # failsafe breaking out after too much iterations, for testing
    iter += 1
    # if iter == 1000:
        # break

iter, registers
```

# --- Day 24: Air Duct Spelunking ---

```python
puzzle = '''###########
#0.1.....2#
#.#######.#
#4.......3#
###########'''

with open('24.txt') as f:
    puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = puzzle[1 : -1]
puzzle = [ line[1 : -1] for line in puzzle ]
puzzle = list(map(list, puzzle))
puzzle

################################################################################
# PART 1-2


# map representation of the puzzle
G = {}
for i in range(len(puzzle)):
    for j in range(len(puzzle[0])):
        if puzzle[i][j] != '#':
            G[ (i, j) ] = puzzle[i][j]
G



def Dijkstra(start):
    global G
    R = max([ node[0] for node in G ])
    C = max([ node[1] for node in G ])

    nodes = { node : { 'checked' : False, 'distance' : 1000000 } for node in G }
    nodes[start]['distance'] = 0

    checking = [start]
    while len(checking) != 0:

        current = checking[0]
        nodes[current]['checked'] = True
        checking.remove(current)

        neighbours = [ (current[0] - 1, current[1]), (current[0] + 1, current[1]), (current[0], current[1] - 1), (current[0], current[1] + 1) ]
        neighbours = [ node for node in neighbours if node in G ]
        for n in neighbours:
            if not nodes[n]['checked'] and n not in checking:
                checking.append(n)
            d = nodes[current]['distance'] + 1
            if d <= nodes[n]['distance']:
                nodes[n]['distance'] = d

    return nodes



target_values = list(range(8))
target_values = list(map(str, target_values))

start_nodes = [ node for node in G if G[node] in target_values ]
start_nodes

distance_maps = { node : Dijkstra(node) for node in start_nodes }
distance_maps = { start_node : { node : distance_maps[start_node][node]['distance'] for node in distance_maps[start_node] } for start_node in distance_maps }
distance_maps = { start_node : { node : distance_maps[start_node][node] for node in distance_maps[start_node] if node != start_node and node in start_nodes } for start_node in distance_maps }
distance_maps

from itertools import permutations
perms = list(permutations(start_nodes))

start = [ node for node in G if G[node] == '0' ][0]
perms = [ p for p in perms if p[0] == start ]

# part 2 tweak, add 0th node to the end
perms = list(map(list, perms))
perms = [ p + [start] for p in perms ]
perms



shortest = 100000
for p in perms:

    score = 0
    for i in range(len(p) - 1):
        score += distance_maps[ p[i] ][ p[i + 1] ]

    if score < shortest:
        shortest = score
        print('Best permutation so far:', p, 'with score:', score)
```

# --- Day 25: Clock Signal ---

```python
# --------- 1.  This section results in the following register values:
# ---------     a = X, b = 0, c = 0, d = X + 2538
# ---------     the algorithm never returns this section
# cpy a d
# cpy 9 c
# cpy 282 b
# inc d
# dec b
# jnz b -2
# dec c
# jnz c -5
# --------- 2.  This section results in the following register values:
# ---------     a = X // 2, b = X % 2, c = X % 2, d = X + 2538
# cpy d a
# jnz 0 0
# cpy a b
# cpy 0 a
# cpy 2 c
# jnz b 2
# jnz 1 6
# dec b
# dec c
# jnz c -4
# inc a
# jnz 1 -7
# --------- 3.  This section results in the following register values:
# ---------     a = X // 2, b = 0, c = 0, d = X + 2538
# cpy 2 b
# jnz c 2
# jnz 1 4
# dec b
# dec c
# jnz 1 -4
# jnz 0 0
# out b         prints out X % 2
# jnz a -19     if a != 0: continue halving and printing
# jnz 1 -21     else a = 2538 + X, starting again

# So we need the lowest input X where:
# X + 2538 is even
# (X + 2538) // 2 is odd
# (X + 2538) // 2 // 2 is even
# (X + 2538) // 2 // 2 // 2 is odd
# ...

# reverse engineering
# 0 => 0 -- from this point, the algo restarts
# 1 => 1
# 2 => 0
# 5 => 1
# 10 => 0
# ...

s = 0
while s < 2538:
    if s % 2 == 0:
        s *= 2
        s += 1
    else:
        s *= 2
        
s - 2538
```
