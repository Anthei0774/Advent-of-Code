# Intcode interpreter

```python
class Intcode:

    #
    def __init__(self, text):
        self.original_memory = text.split(',')
        self.original_memory = list(map(int, self.original_memory))

        self.index = 0
        self.memory = self.original_memory.copy()
        self.halting = False

        
    #
    def __repr__(self):
        return repr(self.memory)

    #
    def step(self):
        # print('Pointer at:', self.index)
        op = self.memory[self.index]

        if op == 99:
            self.halting = True
            return

        input_1_address = self.memory[self.index + 1]
        input_2_address = self.memory[self.index + 2]
        output_address = self.memory[self.index + 3]

        if op == 1:
            # print('Addition')
            self.memory[output_address] = self.memory[input_1_address] + self.memory[input_2_address]
        elif op == 2:
            # print('Multiplication')
            self.memory[output_address] = self.memory[input_1_address] * self.memory[input_2_address]
        else:
            pass

        self.index += 4
        
    #
    def run(self):
        while not self.halting:
            self.step()

    def reset_memory(self):
        self.memory = self.original_memory.copy()

    def reset_index(self):
        self.index = 0

    def full_reset(self):
        self.reset_memory()
        self.reset_index()
        self.halting = False
```

# --- Day 3: Crossed Wires ---

```python
puzzle = '''R8,U5,L5,D3
U7,R6,D4,L4'''
puzzle = '''R75,D30,R83,U83,L12,D49,R71,U7,L72
U62,R66,U55,R34,D71,R55,D58,R83'''
puzzle = '''R98,U47,R26,D63,R33,U87,L62,D20,R33,U53,R51
U98,R91,D20,R16,D67,R40,U7,R15,U6,R7'''
with open('3.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ step.split(',') for step in puzzle ]
puzzle

################################################################################
# PART 1-2

grid = {}

for i in range(2):
    delay = 0

    position = [0, 0]
    for j in range(len(puzzle[i])):
        step = puzzle[i][j]
        direction = None

        if step[0] == 'L': direction = [0, -1]
        if step[0] == 'R': direction = [0, 1]
        if step[0] == 'U': direction = [-1, 0]
        if step[0] == 'D': direction = [1, 0]

        for k in range(int(step[1:])):
            delay += 1

            position[0] += direction[0]
            position[1] += direction[1]

            position = tuple(position)
            if position not in grid:
                grid[position] = { 'wire' : i,
                                   'crossing' : False,
                                   'manhattan' : abs(position[0]) + abs(position[1]),
                                   'delay' : delay}
            else:
                if i != grid[position]['wire']: # only counting collisions with other wire
                    grid[position]['crossing'] = True
                    grid[position]['delay'] += delay
                
            position = list(position)

grid = { position : grid[position] for position in grid if grid[position]['crossing'] }
grid

min( grid[position]['manhattan'] for position in grid if grid[position]['crossing'] ) # part 1

min( grid[position]['delay'] for position in grid if grid[position]['crossing'] ) # part 2
```

# --- Day 6: Universal Orbit Map ---

```python
puzzle = '''COM)B
B)C
C)D
D)E
E)F
B)G
G)H
D)I
E)J
J)K
K)L'''
puzzle += '''
K)YOU
I)SAN'''
with open('6.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ objectPair.split(')') for objectPair in puzzle ]
puzzle

################################################################################
# PART 1

orbits = {}
for orb in puzzle:
    if orb[0] not in orbits:
        orbits[ orb[0] ] = { 'children' : [ orb[1] ]}
    else:
        orbits[ orb[0] ]['children'].append( orb[1] )
    if orb[1] not in orbits:
        orbits[ orb[1] ] = { 'children' : [] }
orbits


currentOrbits = ['COM']
d = 0
s = 0
while currentOrbits != []:

    s += d * len(currentOrbits)

    nextOrbits = []
    for orb in currentOrbits:
        orbits[orb]['depth'] = d
        nextOrbits += orbits[orb]['children']
    
    currentOrbits = nextOrbits.copy()
    d += 1

s

################################################################################
# PART 2

def allDescendants(o):
    global orbits
    D = orbits[o]['children'].copy()
    for c in orbits[o]['children']:
        D += allDescendants(c)
    return D


# 
current = 'YOU'
while True:
    parent = [ orb for orb in orbits if current in orbits[orb]['children'] ]
    assert len(parent) == 1

    parent = parent[0]
    children = allDescendants(parent)
    if 'SAN' in children:
        break
    else:
        current = parent

orbits['YOU']['depth'] + orbits['SAN']['depth'] - 2 * orbits[parent]['depth'] - 2
```

```python
# part 1 inputs
puzzle = '3,15,3,16,1002,16,10,16,1,16,15,15,4,15,99,0,0'
puzzle = '3,23,3,24,1002,24,10,24,1002,23,-1,23,101,5,23,23,1,24,23,23,4,23,99,0,0'
puzzle = '3,31,3,32,1002,32,10,32,1001,31,-2,31,1007,31,0,33,1002,33,7,33,1,33,31,31,1,32,31,31,4,31,99,0,0,0'

# part 2 inputs
puzzle = '3,26,1001,26,-4,26,3,27,1002,27,2,27,1,27,26,27,4,27,1001,28,-1,28,1005,28,6,99,0,0,5'
puzzle = '3,52,1001,52,-5,52,3,53,1,52,56,54,1007,54,5,55,1005,55,26,1001,54,-5,54,1105,1,12,1,53,54,53,1008,54,0,55,1001,55,1,55,2,53,55,53,4,53,1001,56,-1,56,1005,56,6,99,0,0,0,0,10'
with open('7.txt') as f: puzzle = f.read()

puzzle = puzzle.split(',')
puzzle = list(map(int, puzzle))
puzzle

################################################################################
# PART 1

from itertools import permutations
perms = permutations(range(5))
perms = [ p for p in perms ]
perms


S = 0
P = None
for p in perms:
    signal = 0
    

    for phaseSetting in p:
        usePhaseSetting = True

        # running intcode-program here
        memory = puzzle.copy()
        i = 0
        while True:
            if memory[i] == 99:
                break

            cmd = memory[i]
            p1_mode = cmd % 1000 // 100
            p2_mode = cmd // 1000
            # print(i, cmd)

            if cmd % 100 in [1, 2]: # add, multiply    
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]
                
                dst = memory[i + 3]

                if cmd % 100 == 1: memory[ dst ] = p1 + p2
                if cmd % 100 == 2: memory[ dst ] = p1 * p2

                i += 4
                continue

            if cmd % 100 in [3, 4]: # input, output
                dst = memory[i + 1]

                if cmd % 100 == 3:
                    if usePhaseSetting:
                        memory[dst] = phaseSetting
                        usePhaseSetting = False # use phase settings only once
                    else:
                        memory[dst] = signal
                        usePhaseSetting = True # reset for the next thruster
                if cmd % 100 == 4: signal = memory[dst]

                i += 2
                continue

            if cmd % 100 in [5, 6]: # jump-if-true, jump-if-false
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]

                if cmd % 100 == 5: i = p2 if p1 != 0 else i + 3 
                if cmd % 100 == 6: i = p2 if p1 == 0 else i + 3 
                continue

            if cmd % 100 in [7, 8]: # less than, equals
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]
                dst = memory[i + 3]
                if cmd % 100 == 7: memory[dst] = p1 < p2
                if cmd % 100 == 8: memory[dst] = p1 == p2
                i += 4
                continue

    if S < signal:
        S = signal
        P = p
S, P

################################################################################
# PART 2

from time import sleep

perms = permutations(range(5, 10))
perms = [ p for p in perms ]
perms


S = 0
P = None
for p in perms:
    # print('Current permutation:', p)

    # init thrusters for each permutation, each has its on phase setting, index and halting indicator
    thrusters = {
        'A': { 'memory' : puzzle.copy(), 'phaseSetting' : p[0], 'usePhaseSetting' : True, 'i' : 0, 'next' : 'B', 'halting' : False },
        'B': { 'memory' : puzzle.copy(), 'phaseSetting' : p[1], 'usePhaseSetting' : True, 'i' : 0, 'next' : 'C', 'halting' : False },
        'C': { 'memory' : puzzle.copy(), 'phaseSetting' : p[2], 'usePhaseSetting' : True, 'i' : 0, 'next' : 'D', 'halting' : False },
        'D': { 'memory' : puzzle.copy(), 'phaseSetting' : p[3], 'usePhaseSetting' : True, 'i' : 0, 'next' : 'E', 'halting' : False },
        'E': { 'memory' : puzzle.copy(), 'phaseSetting' : p[4], 'usePhaseSetting' : True, 'i' : 0, 'next' : 'A', 'halting' : False }
    }

    currentThruster = 'A'
    i = 0
    memory = puzzle.copy()

    signal = 0

    # feedback loop
    while True:
        if thrusters['E']['halting']: break

        # thruster loop
        while True:
            if thrusters[currentThruster]['halting']: break

            if memory[i] == 99:
                thrusters[currentThruster]['halting'] = True
                # print('Thruster', currentThruster, 'is now halting.')
                break

            cmd = memory[i]
            p1_mode = cmd % 1000 // 100
            p2_mode = cmd // 1000
            # print(i, cmd)

            if cmd % 100 in [1, 2]: # add, multiply    
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]
                
                dst = memory[i + 3]

                if cmd % 100 == 1: memory[ dst ] = p1 + p2
                if cmd % 100 == 2: memory[ dst ] = p1 * p2

                i += 4
                continue

            # modifying input-output handling due to feedback loop
            if cmd % 100 in [3, 4]: # input, output
                dst = memory[i + 1]

                if cmd % 100 == 3:
                    if thrusters[currentThruster]['usePhaseSetting']: # use phase settings only once
                        thrusters[currentThruster]['usePhaseSetting'] = False
                        memory[dst] = thrusters[currentThruster]['phaseSetting']
                    else:
                        memory[dst] = signal # pass signal left and right
                    
                    i += 2
                    continue

                if cmd % 100 == 4:

                    # save signal and memory
                    signal = memory[dst]
                    thrusters[currentThruster]['memory'] = memory.copy()

                    # step with index, save and break out
                    i += 2
                    thrusters[currentThruster]['i'] = i
                    break
        

            if cmd % 100 in [5, 6]: # jump-if-true, jump-if-false
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]

                if cmd % 100 == 5:
                    i = p2 if p1 != 0 else i + 3 
                if cmd % 100 == 6: i = p2 if p1 == 0 else i + 3 
                continue

            if cmd % 100 in [7, 8]: # less than, equals
                p1 = memory[i + 1] if p1_mode else memory[ memory[i + 1] ]
                p2 = memory[i + 2] if p2_mode else memory[ memory[i + 2] ]
                dst = memory[i + 3]
                if cmd % 100 == 7: memory[dst] = p1 < p2
                if cmd % 100 == 8: memory[dst] = p1 == p2
                i += 4
                continue
        
        # if breaking out of the the single thruster loop
        # switch thruster and reset
        currentThruster = thrusters[currentThruster]['next']
        i = thrusters[currentThruster]['i']
        memory = thrusters[currentThruster]['memory'].copy()

    if S < signal:
        S = signal
        P = p

S, P
```

```python
puzzle = '''.#..#
.....
#####
....#
...##'''

puzzle = '''......#.#.
#..#.#....
..#######.
.#.#.###..
.#..#.....
..#....#.#
#..#....#.
.##.#..###
##...#..#.
.#....####'''

puzzle = '''#.#...#.#.
.###....#.
.#....#...
##.#.#.#.#
....#.#.#.
.##..###.#
..#...##..
..##....##
......#...
.####.###.'''

puzzle = '''.#..#..###
####.###.#
....###.#.
..###.##.#
##.##.#.#.
....###..#
..#.#..#.#
#..#.#.###
.##...##.#
.....#.#..'''

puzzle = '''.#..##.###...#######
##.############..##.
.#.######.########.#
.###.#######.####.#.
#####.##.#.##.###.##
..#####..#.#########
####################
#.####....###.#.#.##
##.#################
#####.##.###..####..
..######..##.#######
####.##.####...##..#
.#####..#.######.###
##...#.##########...
#.##########.#######
.####.#.###.###.#.##
....##.##.###..#####
.#.#.###########.###
#.#.#.#####.####.###
###.##.####.##.#..##'''

puzzle = '''.#....#####...#..
##...##.#####..##
##...#...#.#####.
..#.....#...###..
..#.#.....#....##'''

with open('10.txt') as f: puzzle = f.read()

puzzle

################################################################################
# PART 1

import numpy as np

asteroids = {}
r = 0
c = 0
for char in puzzle:
    if char in ['.', '#']:
        if char == '#': asteroids[ (r, c) ] = []
        c += 1
        continue
    if char == '\n':
        r += 1
        c = 0
        continue
asteroids


#
from math import gcd


for i, a1 in enumerate(asteroids):
    for j, a2 in enumerate(asteroids):
        if i + 1 <= j:

            # direction from a1 to a2
            dir = (a2[0] - a1[0], a2[1] - a1[1])

            # normalization
            s = gcd(dir[0], dir[1])
            dir = (dir[0] // s, dir[1] // s)
        
            pos = a1
            while True:
                pos = (pos[0] + dir[0], pos[1] + dir[1])

                if pos in asteroids:
                    break

            if pos == a2:
                asteroids[a1].append(a2)
                asteroids[a2].append(a1)


R = max([ len(asteroids[a]) for a in asteroids ])
A = [ a for a in asteroids if len(asteroids[a]) == R ]
assert len(A) == 1
A = A[0]
A

################################################################################
# PART 2

from math import hypot, atan2, pi


asteroids[A]

# https://stackoverflow.com/questions/41855695/sorting-list-of-two-dimensional-coordinates-by-clockwise-angle-using-python
def clockwise_angl_dist(a):
    global A

    ref = (-1, 0)

    # vector between point and the origin: v = a - A
    # get length and normalize
    v = (a[0] - A[0], a[1] - A[1])
    l = hypot(v[0], v[1])
    v_n = (v[0] / l, v[1] / l)
    

    dotprod  = v_n[0]*ref[0] + v_n[1]*ref[1]     # x1*x2 + y1*y2
    diffprod = ref[1]*v_n[0] - ref[0]*v_n[1]     # x1*y2 - y1*x2
    angl = atan2(diffprod, dotprod)
    # Negative angles represent counter-clockwise angles so we need to subtract them 
    # from 2*pi (360 degrees)
    if angl < 0:
        return 2 * pi + angl, l
    # I return first the angle because that's the primary sorting criterium
    # but if two vectors have the same angle then the shorter distance should come first.
    return angl, l

asteroids[A] = sorted(asteroids[A], key = clockwise_angl_dist)
assert len(asteroids[A]) >= 200

A = asteroids[A][200 - 1]
A[1] * 100 + A[0]
```

```python
with open('11.txt') as f: puzzle = f.read()
puzzle = puzzle.split(',')
puzzle = { i : int(integer) for i, integer in enumerate(puzzle) }
puzzle

################################################################################
# PART 1

# code copied from day 9, modifying outputs

memory = puzzle.copy()
b = 0
i = 0
halting = False

iter = 0
pos = (0, 0)
dir = (0, 1)
activePanels = []
L = 0

while not halting:

    paintOutput = None
    turnOutput = None

    while True:
        if halting: break
        if memory[i] == 99:
            halting = True
            break
        cmd = memory[i]
        # print(i, cmd)

        p1_mode = cmd % 1000 // 100
        p2_mode = cmd % 10000 // 1000
        p3_mode = cmd // 10000
        assert p3_mode != 1

        if cmd % 100 in [1, 2]: # add, multiply
            if p1_mode == 0: p1 = 0 if memory[i + 1] not in memory else memory[ memory[i + 1] ]
            if p1_mode == 1: p1 = memory[i + 1]
            if p1_mode == 2: p1 = 0 if b + memory[i + 1] not in memory else memory[b + memory[i + 1]]

            if p2_mode == 0: p2 = 0 if memory[i + 2] not in memory else memory[ memory[i + 2] ]
            if p2_mode == 1: p2 = memory[i + 2]
            if p2_mode == 2: p2 = 0 if b + memory[i + 2] not in memory else memory[b + memory[i + 2]]
            
            if p3_mode == 0: dst = memory[i + 3]
            if p3_mode == 2: dst = b + memory[i + 3]

            if cmd % 100 == 1: memory[dst] = p1 + p2
            if cmd % 100 == 2: memory[dst] = p1 * p2

            i += 4
            continue


        if cmd % 100 in [3, 4, 9]: # input, output, adjust relative base

            if cmd % 100 == 3:
                if p1_mode == 0: dst = memory[i + 1]
                if p1_mode == 2: dst = b + memory[i + 1] 
                memory[dst] = 1 if pos in activePanels else 0 # provide 0 if the robot is over a black panel or 1 if the robot is over a white panel
                i += 2
                continue

            if cmd % 100 in [4, 9]:
                if p1_mode == 0: val = 0 if memory[i + 1] not in memory else memory[ memory[i + 1] ]
                if p1_mode == 1: val = memory[i + 1]
                if p1_mode == 2: val = 0 if b + memory[i + 1] not in memory else memory[b + memory[i + 1]]
                
                if cmd % 100 == 4:
                    i += 2
                    if paintOutput == None:
                        paintOutput = val
                    else:
                        turnOutput = val
                        break

                if cmd % 100 == 9:
                    b += val
                    i += 2
                    continue

            

        if cmd % 100 in [5, 6]: # jump-if-true, jump-if-false
            if p1_mode == 0: p1 = 0 if memory[i + 1] not in memory else memory[ memory[i + 1] ]
            if p1_mode == 1: p1 = memory[i + 1]
            if p1_mode == 2: p1 = 0 if b + memory[i + 1] not in memory else memory[b + memory[i + 1]]

            if p2_mode == 0: p2 = 0 if memory[i + 2] not in memory else memory[ memory[i + 2] ]
            if p2_mode == 1: p2 = memory[i + 2]
            if p2_mode == 2: p2 = 0 if b + memory[i + 2] not in memory else memory[b + memory[i + 2]]

            if cmd % 100 == 5: i = p2 if p1 != 0 else i + 3 
            if cmd % 100 == 6: i = p2 if p1 == 0 else i + 3 
            continue

        if cmd % 100 in [7, 8]: # less than, equals
            if p1_mode == 0: p1 = 0 if memory[i + 1] not in memory else memory[ memory[i + 1] ]
            if p1_mode == 1: p1 = memory[i + 1]
            if p1_mode == 2: p1 = 0 if b + memory[i + 1] not in memory else memory[b + memory[i + 1]]

            if p2_mode == 0: p2 = 0 if memory[i + 2] not in memory else memory[ memory[i + 2] ]
            if p2_mode == 1: p2 = memory[i + 2]
            if p2_mode == 2: p2 = 0 if b + memory[i + 2] not in memory else memory[b + memory[i + 2]]
            
            if p3_mode == 0: dst = memory[i + 3]
            if p3_mode == 2: dst = b + memory[i + 3]

            if cmd % 100 == 7: memory[dst] = p1 < p2
            if cmd % 100 == 8: memory[dst] = p1 == p2
            i += 4
            continue

    iter += 1
    paintOutput = '#' if paintOutput == 0 else '.'
    turnOutput = 'L' if turnOutput == 0 else 'R'
    print('Iter:', iter, '; paint:', paintOutput, '; turn:', turnOutput)

    if paintOutput == '#':
        if pos not in activePanels:
            activePanels.append(pos)
            l = len(activePanels)
            if L < l:
                L = l
    else:
        if pos in activePanels:
            activePanels.remove(pos)

    if turnOutput == 'L':
        dir = (-dir[1], dir[0])
    else:
        dir = (dir[1], -dir[0])

    pos = (pos[0] + dir[0], pos[1] + dir[1])
    

L
```

```python
puzzle = '''         A           
         A           
  #######.#########  
  #######.........#  
  #######.#######.#  
  #######.#######.#  
  #######.#######.#  
  #####  B    ###.#  
BC...##  C    ###.#  
  ##.##       ###.#  
  ##...DE  F  ###.#  
  #####    G  ###.#  
  #########.#####.#  
DE..#######...###.#  
  #.#########.###.#  
FG..#########.....#  
  ###########.#####  
             Z       
             Z       '''

puzzle = '''                   A               
                   A               
  #################.#############  
  #.#...#...................#.#.#  
  #.#.#.###.###.###.#########.#.#  
  #.#.#.......#...#.....#.#.#...#  
  #.#########.###.#####.#.#.###.#  
  #.............#.#.....#.......#  
  ###.###########.###.#####.#.#.#  
  #.....#        A   C    #.#.#.#  
  #######        S   P    #####.#  
  #.#...#                 #......VT
  #.#.#.#                 #.#####  
  #...#.#               YN....#.#  
  #.###.#                 #####.#  
DI....#.#                 #.....#  
  #####.#                 #.###.#  
ZZ......#               QG....#..AS
  ###.###                 #######  
JO..#.#.#                 #.....#  
  #.#.#.#                 ###.#.#  
  #...#..DI             BU....#..LF
  #####.#                 #.#####  
YN......#               VT..#....QG
  #.###.#                 #.###.#  
  #.#...#                 #.....#  
  ###.###    J L     J    #.#.###  
  #.....#    O F     P    #.#...#  
  #.###.#####.#.#####.#####.###.#  
  #...#.#.#...#.....#.....#.#...#  
  #.#####.###.###.#.#.#########.#  
  #...#.#.....#...#.#.#.#.....#.#  
  #.###.#####.###.###.#.#.#######  
  #.#.........#...#.............#  
  #########.###.###.#############  
           B   J   C               
           U   P   P               '''

with open('20.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = list(map(list, puzzle))
# for row in puzzle: print(row)

################################################################################
# PART 1

R = len(puzzle)
C = len(puzzle[0])

graph = {}
portals = {}

for r in range(R):
    for c in range(C):

        tile = puzzle[r][c]
        if tile == '.':

            graph[ (r, c) ] = []

            neighbours = [ (r - 1, c), (r + 1, c), (r, c + 1), (r, c - 1) ]
            neighbours = [ n for n in neighbours if puzzle[ n[0] ][ n[1] ] != '#' ] # any which are not wall

            while len(neighbours) != 0:

                n = neighbours.pop()

                if puzzle[ n[0] ][ n[1] ] == '.': # open passage
                    graph[ (r, c) ].append(n)

                else: # a portal
                    d = (n[0] - r, n[1] - c)
                    
                    # check next char in direction => get portal name
                    if d == (-1, 0):
                        p = puzzle[ n[0] ][ n[1] ] + puzzle[ n[0] - 1 ][ n[1] ]
                    elif d == (1, 0):
                        p = puzzle[ n[0] ][ n[1] ] + puzzle[ n[0] + 1 ][ n[1] ]
                    elif d == (0, 1):
                        p = puzzle[ n[0] ][ n[1] ] + puzzle[ n[0] ][ n[1] + 1 ]
                    else:
                        assert d == (0, -1)
                        p = puzzle[ n[0] ][ n[1] ] + puzzle[ n[0] ][ n[1] - 1 ]

                    p = sorted(p)
                    p = ''.join(p)
                    
                    # connect portals
                    if p not in portals:
                        portals[p] = [ (r, c) ]
                    else:
                        portals[p].append( (r, c) )

# connect portals in graph
for p in portals:
    if len(portals[p]) == 2:
        graph[ portals[p][0] ].append( portals[p][1] )
        graph[ portals[p][1] ].append( portals[p][0] )

graph


# Dijkstra
start = portals['AA'][0]
end = portals['ZZ'][0]
distances = { node : 1000000 for node in graph }
distances[start] = 0
distances

buffer = [ start ]
while len(buffer) != 0:

    curr = buffer.pop()
    neighbours = graph[curr]

    for next in neighbours:
        if distances[curr] + 1 < distances[next]:
            buffer.append(next)
            distances[next] = distances[curr] + 1

distances[end]
```
