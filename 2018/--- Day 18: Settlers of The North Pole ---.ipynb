```python
puzzle = '''.#.#...|#.
.....#|##|
.|..|...#.
..|#.....#
#.#|||#|#|
...#.||...
.|....|...
||...#|.#|
|.||||..|.
...#.|..|.'''
with open('18.txt') as f: puzzle = f.read()

puzzle = puzzle.split('\n')
puzzle = [ list(line) for line in puzzle ]
puzzle

R = len(puzzle)
C = len(puzzle[0])

################################################################################
# PART 1

def calculate(current):
    global R, C

    next = [ line.copy() for line in current ]
    for r in range(R):
        for c in range(C):

            neighbours = [ (r + i, c + j) for i in [-1, 0, 1] for j in [-1, 0, 1] ]
            neighbours = [ n for n in neighbours if 0 <= n[0] and n[0] < R and 0 <= n[1] and n[1] < C and n != (r, c) ]
            neighbours = [ current[ n[0] ][ n[1] ] for n in neighbours ]

            # An open acre will become filled with trees if three or more adjacent acres contained trees. Otherwise, nothing happens.
            if current[r][c] == '.' and neighbours.count('|') >= 3:
                next[r][c] = '|'
                continue
            # An acre filled with trees will become a lumberyard if three or more adjacent acres were lumberyards. Otherwise, nothing happens.
            if current[r][c] == '|' and neighbours.count('#') >= 3:
                next[r][c] = '#'
                continue
            # An acre containing a lumberyard will remain a lumberyard if it was adjacent to at least one other lumberyard and at least one acre containing trees. Otherwise, it becomes open.
            if current[r][c] == '#':
                if neighbours.count('|') >= 1 and neighbours.count('#') >= 1:
                    next[r][c] = '#'
                else:
                    next[r][c] = '.'
                continue

    return next


current = [ line.copy() for line in puzzle ]
for iter in range(10):
    next = calculate(current)
    current = [ line.copy() for line in next ]

# total resource value
W = sum(line.count('|') for line in current)
L = sum(line.count('#') for line in current)
W * L

################################################################################
# PART 2

def encode_to_state(current):
    s = [ ''.join(line) for line in current ]
    s = ''.join(s)
    return s

# save states ==> expecting cyclical pattern
current = [ line.copy() for line in puzzle ]
states = [ encode_to_state(current) ]
states

#
I = 1000000000
for iter in range(I):
    next = calculate(current)

    # check if state is already encountered
    s = encode_to_state(next)
    if s in states:
        i = states.index(s)
        break
    else:
        states.append(s)

    current = [ line.copy() for line in next ]

# to get the final state index
# 1. subtract num of states until cycle starts
# 2. take mod cycle
I -= i
I %= (len(states) - i)
I

# i + I marks the final state
W = states[i + I].count('|')
L = states[i + I].count('#')
W * L
```
