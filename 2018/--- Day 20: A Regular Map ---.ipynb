```python
puzzle = '''^ENWWW(NEEE|SSE(EE|N))$'''
puzzle = '''^ENNWSWW(NEWS|)SSSEEN(WNSE|)EE(SWEN|)NNN$'''
puzzle = '''^ESSWWN(E|NNENN(EESS(WNSE|)SSS|WWWSSSSE(SW|NNNE)))$'''
puzzle = '''^WSSEESWWWNW(S|NENNEEEENN(ESSSSW(NWSW|SSEN)|WSWWN(E|WWS(E|SS))))$'''
with open('20.txt') as f: puzzle = f.read()

puzzle = puzzle[1:-1]
puzzle

################################################################################
# PART 1-2

walk = { 0 : (0, 0) }

graph = { (0, 0) : { 'neighbours' : [], 'type' : '.', 'dist' : 0 } }


l = 0
i = 0
while i < len(puzzle):

    assert l >= 0, 'Level cannot be negative'

    p = walk[l]
    assert p in graph, 'Position must in graph'

    c = puzzle[i]
    # print('Char:', c)

    if c != '|':

        # char is a letter E, W, N, S --- step in direction
        if c in ['E', 'W', 'N', 'S']:

            # next position
            n = None
            if c == 'E':
                n = (p[0], p[1] + 2)
            elif c == 'W':
                n = (p[0], p[1] - 2)
            elif c == 'N':
                n = (p[0] - 2, p[1])
            else:
                assert c == 'S'
                n = (p[0] + 2, p[1])
            # print('Next:', n)

            # position between current and next
            m = (p[0] + n[0], p[1] + n[1])
            m = (m[0] // 2, m[1] // 2)
            # print('Mid:', m)

            t = None
            if m[0] % 2 == 0:
                t = '|'
            else:
                assert m[1] % 2 == 0
                t = '-'
            # print('Type:', t)

            if m not in graph:
                graph[m] = { 'neighbours' : [p, n], 'type' : t }

            if m not in graph[p]['neighbours']:
                graph[p]['neighbours'].append(m)

            if n not in graph:
                graph[n] = { 'neighbours' : [m], 'type' : '.', 'dist' : graph[p]['dist'] + 1 }

        # char is an opening bracket ( --- go a level deeper
        elif c == '(':
            l += 1

        # char is a closing bracket ) --- destroy current level
        else:
            assert c == ')'
            del walk[l]
            l -= 1

        # whatever adjustments were made; in the end, set the new position
        walk[l] = n


    else:
        assert l > 0, 'Level must be positive when encountering an OR part'
        walk[l] = walk[l - 1]

    # print('Walk:', walk, '\n')
    i += 1

graph

# PART 1
max( graph[p]['dist'] for p in graph if graph[p]['type'] == '.' )

# PART 2
tmp = { p : graph[p] for p in graph if graph[p]['type'] == '.' }
tmp = { p : tmp[p] for p in tmp if tmp[p]['dist'] >= 1000 }
len(tmp)
```

```python
def draw(graph):

    m_x = min(p[0] for p in graph) - 1
    M_x = max(p[0] for p in graph) + 2
    m_y = min(p[1] for p in graph) - 1
    M_y = max(p[1] for p in graph) + 2

    image = ''
    for x in range(m_x, M_x):
        for y in range(m_y, M_y):
            p = (x, y)
            if p not in graph:
                image += '#'
            else:
                image += graph[p]['type']
        image += '\n'

    print(image)

draw(graph)
```
