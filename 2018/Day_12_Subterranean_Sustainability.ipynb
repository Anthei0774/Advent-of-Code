```python
puzzle = '''initial state: #..#.#..##......###...###

...## => #
..#.. => #
.#... => #
.#.#. => #
.#.## => #
.##.. => #
.#### => #
#.#.# => #
#.### => #
##.#. => #
##.## => #
###.. => #
###.# => #
####. => #'''
with open('12.txt') as f: puzzle = f.read()

#
puzzle = puzzle.split('\n')

#
grow_patterns = puzzle[2:]
grow_patterns = [ p for p in grow_patterns if p[-1] == '#' ]
grow_patterns = [ p[: -5] for p in grow_patterns]
grow_patterns

#
pots = puzzle[0]
pots = pots[15:]
pots

################################################################################
# PART 1

# storing only indices of grown plants
grown_pots = [ i for (i, p) in enumerate(pots) if p == '#' ]
grown_pots

for iter in range(20):
    # print(iter, '---', grown_pots)

    # neighbours to be checked, because they have a grown pot nearby
    empty_neighbours = []
    for p in grown_pots:
        for i in [-4, -3, -2, -1, 1, 2, 3, 4]: # 8 adjacent tiles
            n = p + i
            if n not in grown_pots:
                empty_neighbours.append(n)

    # deduplication, sorting and dropping min-2 and max-2
    # as growing counts only the two nearest neighbours
    empty_neighbours = sorted(set(empty_neighbours))
    empty_neighbours = empty_neighbours[2:-2]
    # print(iter, '---', empty_neighbours)

    # main container to be filled
    next_iteration = []

    # first checking surviving pots
    for p in grown_pots:
        pattern = ''
        for i in [-2, -1, 0, 1, 2]: # 4 adjacent tiles
            n = p + i
            pattern += '#' if n in grown_pots else '.'
        if pattern in grow_patterns:
            next_iteration.append(p)

    # same with neighbours pots, plant will grow there
    for n in empty_neighbours:
        pattern = ''
        for i in [-2, -1, 0, 1, 2]: # 4 adjacent tiles
            nn = n + i
            pattern += '#' if nn in grown_pots else '.'
        if pattern in grow_patterns:
            next_iteration.append(n)

    next_iteration = sorted(next_iteration)
    # print(iter, '---', next_iteration)

    # iter step
    grown_pots = next_iteration.copy()

sum(grown_pots)

################################################################################
# PART 2

# solution thanks to
# https://www.reddit.com/r/adventofcode/comments/a5eztl/comment/ebm4c9d/
patience = 100
w = 0
diffs = []

# storing only indices of grown plants
grown_pots = [ i for (i, p) in enumerate(pots) if p == '#' ]
grown_pots
S = sum(grown_pots)

# iter go brrrrrrrrrrr...
I = 50000000000
for iter in range(I):
    # print(iter, '---', grown_pots)

    # neighbours to be checked, because they have a grown pot nearby
    empty_neighbours = []
    for p in grown_pots:
        for i in [-4, -3, -2, -1, 1, 2, 3, 4]: # 8 adjacent tiles
            n = p + i
            if n not in grown_pots:
                empty_neighbours.append(n)

    # deduplication, sorting and dropping min-2 and max-2
    # as growing counts only the two nearest neighbours
    empty_neighbours = sorted(set(empty_neighbours))
    empty_neighbours = empty_neighbours[2:-2]
    # print(iter, '---', empty_neighbours)

    # main container to be filled
    next_iteration = []

    # first checking surviving pots
    for p in grown_pots:
        pattern = ''
        for i in [-2, -1, 0, 1, 2]: # 4 adjacent tiles
            n = p + i
            pattern += '#' if n in grown_pots else '.'
        if pattern in grow_patterns:
            next_iteration.append(p)

    # same with neighbours pots, plant will grow there
    for n in empty_neighbours:
        pattern = ''
        for i in [-2, -1, 0, 1, 2]: # 4 adjacent tiles
            nn = n + i
            pattern += '#' if nn in grown_pots else '.'
        if pattern in grow_patterns:
            next_iteration.append(n)

    next_iteration = sorted(next_iteration)
    # print(iter, '---', next_iteration)

    # pattern recognition for early abort
    s_current = sum(grown_pots)
    s_next = sum(next_iteration)
    d = s_next - s_current
    if iter != 0 and d == diffs[-1]:
        w += 1
        if w == patience: # wait until sure
            # print('Repeating increment:', d)
            # print(diffs[-patience - 5 :])
            break
    else:
        diffs.append(d)
        w = 0
        # print(iter, '---', s_current, s_next, d)

    # iter step
    grown_pots = next_iteration.copy()
    S = sum(grown_pots)



# end, so the final score is
# current pot sum (S) + remaining_time * d
# where remaining time is
# all iterations (I) - flickering diffs (l) - waiting (w)
l = len(diffs)
S + (I - l - w) * d
```
